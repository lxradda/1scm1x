```python
# requirements.txt
web3==6.12.0
pandas==2.0.3
scikit-learn==1.3.0
gradio==3.45.0
requests==2.31.0
python-dotenv==1.0.0
joblib==1.3.2
numpy==1.24.3
tweepy==4.14.0
beautifulsoup4==4.12.2

# .env (Ã‡evresel deÄŸiÅŸkenler)
INFURA_PROJECT_ID=your_infura_project_id
ETHERSCAN_API_KEY=your_etherscan_api_key
TWITTER_BEARER_TOKEN=your_twitter_bearer_token

# config.py
import os
from dotenv import load_dotenv

load_dotenv()

class Config:
    INFURA_PROJECT_ID = os.getenv('INFURA_PROJECT_ID')
    ETHERSCAN_API_KEY = os.getenv('ETHERSCAN_API_KEY')
    TWITTER_BEARER_TOKEN = os.getenv('TWITTER_BEARER_TOKEN')
    INFURA_URL = f"https://mainnet.infura.io/v3/{INFURA_PROJECT_ID}"

# utils/onchain_checker.py
import requests
from web3 import Web3
from config import Config
import json
import time

class OnchainChecker:
    def __init__(self):
        self.w3 = Web3(Web3.HTTPProvider(Config.INFURA_URL))
        self.etherscan_api = Config.ETHERSCAN_API_KEY

    def get_contract_features(self, address):
        try:
            # Kontrat adresini doÄŸrula
            if not self.w3.is_address(address):
                raise ValueError("GeÃ§ersiz kontrat adresi")

            address = self.w3.to_checksum_address(address)

            # Token bilgilerini al
            token_info = self._get_token_info(address)
            holder_data = self._get_holder_distribution(address)
            contract_code = self._get_contract_source(address)
            liquidity_info = self._check_liquidity_lock(address)

            return {
                "holder_concentration": self._calculate_concentration(holder_data),
                "mint_function": self._check_mint_function(contract_code),
                "lp_locked": liquidity_info.get("locked", False),
                "max_tax": self._extract_max_tax(contract_code),
                "owner_renounced": self._check_ownership_renounced(address),
                "contract_verified": contract_code.get("verified", False),
                "creation_date": self._get_creation_date(address),
                "total_supply": token_info.get("total_supply", 0),
                "holder_count": len(holder_data)
            }
        except Exception as e:
            print(f"Hata: {e}")
            return self._get_default_features()

    def _get_token_info(self, address):
        """Token temel bilgilerini al"""
        try:
            # ERC-20 ABI (sadece gerekli fonksiyonlar)
            erc20_abi = [
                {
                    "constant": True,
                    "inputs": [],
                    "name": "totalSupply",
                    "outputs": [{"name": "", "type": "uint256"}],
                    "type": "function"
                }
            ]

            contract = self.w3.eth.contract(address=address, abi=erc20_abi)
            total_supply = contract.functions.totalSupply().call()

            return {"total_supply": total_supply}
        except:
            return {"total_supply": 0}

    def _get_holder_distribution(self, address):
        """Holder daÄŸÄ±lÄ±mÄ±nÄ± Etherscan'den al"""
        try:
            url = f"https://api.etherscan.io/api"
            params = {
                "module": "token",
                "action": "tokenholderlist",
                "contractaddress": address,
                "page": 1,
                "offset": 100,
                "apikey": self.etherscan_api
            }

            response = requests.get(url, params=params, timeout=10)
            data = response.json()

            if data.get("status") == "1":
                return data.get("result", [])
            return []
        except:
            return []

    def _calculate_concentration(self, holders):
        """Top 10 holder konsantrasyonunu hesapla"""
        if not holders:
            return 100  # Veri yoksa risk yÃ¼ksek

        try:
            # Ä°lk 10 holder'Ä±n toplam token yÃ¼zdesi
            top_10_balance = sum(float(h.get("TokenHolderQuantity", 0)) for h in holders[:10])
            total_supply = sum(float(h.get("TokenHolderQuantity", 0)) for h in holders)

            if total_supply == 0:
                return 100

            concentration = (top_10_balance / total_supply) * 100
            return min(concentration, 100)  # Max %100
        except:
            return 75  # Default deÄŸer

    def _get_contract_source(self, address):
        """Kontrat kaynak kodunu al"""
        try:
            url = f"https://api.etherscan.io/api"
            params = {
                "module": "contract",
                "action": "getsourcecode",
                "address": address,
                "apikey": self.etherscan_api
            }

            response = requests.get(url, params=params, timeout=10)
            data = response.json()

            if data.get("status") == "1" and data.get("result"):
                source_code = data["result"][0].get("SourceCode", "")
                return {
                    "source_code": source_code,
                    "verified": len(source_code) > 0
                }
            return {"source_code": "", "verified": False}
        except:
            return {"source_code": "", "verified": False}

    def _check_mint_function(self, contract_data):
        """Mint fonksiyonu var mÄ± kontrol et"""
        source_code = contract_data.get("source_code", "").lower()
        mint_keywords = ["function mint", "mint(", "_mint(", "mintto", "mint_"]
        return any(keyword in source_code for keyword in mint_keywords)

    def _extract_max_tax(self, contract_data):
        """Maksimum tax oranÄ±nÄ± bulmaya Ã§alÄ±ÅŸ"""
        source_code = contract_data.get("source_code", "").lower()

        # YaygÄ±n tax variable isimleri
        tax_patterns = ["buytax", "selltax", "tax", "fee"]

        import re
        for pattern in tax_patterns:
            # uint256 public buyTax = 5; gibi patternleri ara
            regex = rf"{pattern}\s*=\s*(\d+)"
            matches = re.findall(regex, source_code)
            if matches:
                return max(int(match) for match in matches)

        return 10  # Default deÄŸer

    def _check_ownership_renounced(self, address):
        """Owner renounce edilmiÅŸ mi kontrol et"""
        try:
            # Owner fonksiyonu Ã§aÄŸÄ±r
            owner_abi = [{
                "constant": True,
                "inputs": [],
                "name": "owner",
                "outputs": [{"name": "", "type": "address"}],
                "type": "function"
            }]

            contract = self.w3.eth.contract(address=address, abi=owner_abi)
            owner = contract.functions.owner().call()

            # 0x000...000 adresine sahipse renounce edilmiÅŸ
            return owner == "0x0000000000000000000000000000000000000000"
        except:
            return False

    def _check_liquidity_lock(self, address):
        """Likidite kilidi kontrol et (basit versiyon)"""
        # Bu gerÃ§ek implementasyonda Uniswap/PancakeSwap pool kontratlarÄ±nÄ± kontrol etmeli
        return {"locked": False, "lock_time": 0}

    def _get_creation_date(self, address):
        """Kontrat oluÅŸturulma tarihini al"""
        try:
            url = f"https://api.etherscan.io/api"
            params = {
                "module": "account",
                "action": "txlist",
                "address": address,
                "startblock": 0,
                "endblock": 99999999,
                "page": 1,
                "offset": 1,
                "sort": "asc",
                "apikey": self.etherscan_api
            }

            response = requests.get(url, params=params, timeout=10)
            data = response.json()

            if data.get("status") == "1" and data.get("result"):
                timestamp = int(data["result"][0].get("timeStamp", 0))
                return timestamp
            return 0
        except:
            return 0

    def _get_default_features(self):
        """Hata durumunda default deÄŸerler"""
        return {
            "holder_concentration": 75,
            "mint_function": True,
            "lp_locked": False,
            "max_tax": 25,
            "owner_renounced": False,
            "contract_verified": False,
            "creation_date": 0,
            "total_supply": 0,
            "holder_count": 0
        }

# utils/social_analysis.py
import requests
import tweepy
from config import Config
from datetime import datetime, timedelta

class SocialAnalyzer:
    def __init__(self):
        if Config.TWITTER_BEARER_TOKEN:
            self.twitter_client = tweepy.Client(bearer_token=Config.TWITTER_BEARER_TOKEN)
        else:
            self.twitter_client = None

    def analyze_social_signals(self, token_name, contract_address):
        """Token iÃ§in sosyal medya sinyallerini analiz et"""
        signals = {
            "twitter_mentions": 0,
            "sentiment_score": 0.5,  # 0-1 arasÄ±
            "bot_activity": False,
            "pump_dump_keywords": False
        }

        if self.twitter_client:
            try:
                # Son 24 saat iÃ§inde token hakkÄ±nda tweetleri ara
                query = f"{token_name} OR {contract_address[:10]} -is:retweet lang:en"
                tweets = self.twitter_client.search_recent_tweets(
                    query=query,
                    max_results=50,
                    tweet_fields=["created_at", "author_id", "public_metrics"]
                )

                if tweets.data:
                    signals["twitter_mentions"] = len(tweets.data)
                    signals["pump_dump_keywords"] = self._check_pump_dump_keywords(tweets.data)
                    signals["bot_activity"] = self._detect_bot_activity(tweets.data)
                    signals["sentiment_score"] = self._calculate_sentiment(tweets.data)

            except Exception as e:
                print(f"Twitter API hatasÄ±: {e}")

        return signals

    def _check_pump_dump_keywords(self, tweets):
        """Pump&dump anahtar kelimelerini kontrol et"""
        pump_keywords = [
            "moon", "ðŸš€", "pump", "easy money", "quick profit",
            "100x", "1000x", "get rich", "diamond hands", "ape in"
        ]

        tweet_texts = " ".join([tweet.text.lower() for tweet in tweets])
        return any(keyword in tweet_texts for keyword in pump_keywords)

    def _detect_bot_activity(self, tweets):
        """Bot aktivitesi tespit et"""
        if len(tweets) < 10:
            return False

        # AynÄ± metinlerin tekrarÄ±
        tweet_texts = [tweet.text for tweet in tweets]
        unique_tweets = set(tweet_texts)

        # %70'den fazlasÄ± aynÄ±ysa bot aktivitesi olabilir
        return len(unique_tweets) / len(tweet_texts) < 0.3

    def _calculate_sentiment(self, tweets):
        """Basit sentiment analizi"""
        positive_words = ["good", "great", "amazing", "bullish", "buy"]
        negative_words = ["scam", "rug", "fake", "avoid", "dump"]

        total_score = 0
        for tweet in tweets:
            text = tweet.text.lower()
            pos_count = sum(word in text for word in positive_words)
            neg_count = sum(word in text for word in negative_words)

            if pos_count + neg_count > 0:
                total_score += pos_count / (pos_count + neg_count)
            else:
                total_score += 0.5  # NÃ¶tr

        return total_score / len(tweets) if tweets else 0.5

# utils/code_analysis.py
import re

class CodeAnalyzer:
    def __init__(self):
        self.risk_patterns = {
            "mint_function": [
                r"function\s+mint\s*\(",
                r"function\s+_mint\s*\(",
                r"\.mint\s*\(",
                r"mintTo\s*\("
            ],
            "ownership_issues": [
                r"onlyOwner",
                r"_owner\s*=",
                r"transferOwnership",
                r"renounceOwnership"
            ],
            "liquidity_removal": [
                r"removeLiquidity",
                r"withdraw.*Liquidity",
                r"emergencyWithdraw"
            ],
            "tax_functions": [
                r"buyTax\s*=",
                r"sellTax\s*=",
                r"setTax",
                r"updateTax"
            ],
            "blacklist_functions": [
                r"blacklist",
                r"addBot",
                r"removeBot",
                r"setBot"
            ],
            "pause_functions": [
                r"pause\s*\(",
                r"unpause\s*\(",
                r"setPaused"
            ]
        }

    def analyze_contract_code(self, address, source_code=""):
        """Kontrat kodunu analiz ederek risk bayraklarÄ±nÄ± dÃ¶ndÃ¼r"""
        flags = []

        if not source_code:
            flags.append("âš ï¸ Kontrat kodu doÄŸrulanamadÄ±")
            return flags

        source_code = source_code.lower()

        # Mint fonksiyonu kontrolÃ¼
        if self._check_pattern_exists(source_code, self.risk_patterns["mint_function"]):
            flags.append("ðŸ”´ Mint fonksiyonu tespit edildi")

        # Ownership kontrolleri
        if self._check_pattern_exists(source_code, self.risk_patterns["ownership_issues"]):
            if not self._check_ownership_renounced(source_code):
                flags.append("ðŸ”´ Owner haklarÄ± korunuyor")

        # Likidite Ã§ekme fonksiyonlarÄ±
        if self._check_pattern_exists(source_code, self.risk_patterns["liquidity_removal"]):
            flags.append("ðŸ”´ Likidite Ã§ekme fonksiyonu var")

        # Vergi fonksiyonlarÄ±
        if self._check_pattern_exists(source_code, self.risk_patterns["tax_functions"]):
            tax_rate = self._extract_tax_rate(source_code)
            if tax_rate > 10:
                flags.append(f"ðŸ”´ YÃ¼ksek vergi oranÄ±: %{tax_rate}")

        # Blacklist fonksiyonlarÄ±
        if self._check_pattern_exists(source_code, self.risk_patterns["blacklist_functions"]):
            flags.append("ðŸ”´ Blacklist fonksiyonu tespit edildi")

        # Pause fonksiyonlarÄ±
        if self._check_pattern_exists(source_code, self.risk_patterns["pause_functions"]):
            flags.append("ðŸ”´ Trading durdurma fonksiyonu var")

        # Honeypot kontrolleri
        if self._check_honeypot_patterns(source_code):
            flags.append("ðŸ”´ Honeypot kalÄ±plarÄ± tespit edildi")

        if not flags:
            flags.append("âœ… Kod analizi temiz")

        return flags

    def _check_pattern_exists(self, code, patterns):
        """Verilen pattern'lerden herhangi biri var mÄ± kontrol et"""
        for pattern in patterns:
            if re.search(pattern, code, re.IGNORECASE):
                return True
        return False

    def _check_ownership_renounced(self, code):
        """Ownership renounce edilmiÅŸ mi kontrol et"""
        renounce_patterns = [
            r"renounceownership\s*\(\s*\)",
            r"_owner\s*=.*0x0+",
            r"owner.*=.*address\(0\)"
        ]
        return self._check_pattern_exists(code, renounce_patterns)

    def _extract_tax_rate(self, code):
        """Kod iÃ§inden vergi oranÄ±nÄ± Ã§Ä±karma"""
        tax_patterns = [
            r"buytax\s*=\s*(\d+)",
            r"selltax\s*=\s*(\d+)",
            r"tax.*=\s*(\d+)"
        ]

        max_tax = 0
        for pattern in tax_patterns:
            matches = re.findall(pattern, code, re.IGNORECASE)
            for match in matches:
                max_tax = max(max_tax, int(match))

        return max_tax

    def _check_honeypot_patterns(self, code):
        """Honeypot kalÄ±plarÄ±nÄ± kontrol et"""
        honeypot_patterns = [
            r"require.*balances\[.*\]\s*>=",  # Balance kontrolÃ¼
            r"if.*balances\[.*\].*return false",  # Transfer engelleme
            r"mapping.*bool.*isexcluded",  # Exclusion mapping
            r"tradingopen\s*=\s*false"  # Trading kapalÄ±
        ]
        return self._check_pattern_exists(code, honeypot_patterns)

# utils/feature_engineering.py
import numpy as np
from datetime import datetime

class FeatureEngineer:
    def build_features(self, onchain_data, social_data=None):
        """Ham verilerden ML modeli iÃ§in Ã¶zellikler oluÅŸtur"""

        # Temel Ã¶zellikler
        features = [
            onchain_data.get("holder_concentration", 75),
            int(onchain_data.get("mint_function", True)),
            int(onchain_data.get("lp_locked", False)),
            onchain_data.get("max_tax", 25),
            int(onchain_data.get("owner_renounced", False)),
            int(onchain_data.get("contract_verified", False))
        ]

        # Kontrat yaÅŸÄ± (gÃ¼n cinsinden)
        creation_date = onchain_data.get("creation_date", 0)
        if creation_date > 0:
            age_days = (datetime.now().timestamp() - creation_date) / 86400
            features.append(min(age_days, 365))  # Max 1 yÄ±l
        else:
            features.append(0)

        # Holder sayÄ±sÄ± (log Ã¶lÃ§eÄŸi)
        holder_count = onchain_data.get("holder_count", 1)
        features.append(np.log10(max(holder_count, 1)))

        # Sosyal medya Ã¶zellikleri
        if social_data:
            features.extend([
                social_data.get("twitter_mentions", 0),
                social_data.get("sentiment_score", 0.5) * 100,  # 0-100 Ã¶lÃ§eÄŸi
                int(social_data.get("bot_activity", False)),
                int(social_data.get("pump_dump_keywords", False))
            ])
        else:
            features.extend([0, 50, 0, 0])  # Default deÄŸerler

        return np.array(features)

    def get_feature_names(self):
        """Ã–zellik isimlerini dÃ¶ndÃ¼r"""
        return [
            "holder_concentration",
            "mint_function",
            "lp_locked",
            "max_tax",
            "owner_renounced",
            "contract_verified",
            "age_days",
            "log_holder_count",
            "twitter_mentions",
            "sentiment_score",
            "bot_activity",
            "pump_dump_keywords"
        ]
```

```python
# main.py
from utils.onchain_checker import OnchainChecker
from utils.social_analysis import SocialAnalyzer
from utils.code_analysis import CodeAnalyzer
from utils.feature_engineering import FeatureEngineer
import joblib
import os
import numpy as np
from datetime import datetime

class ScamDetector:
    def __init__(self):
        self.onchain_checker = OnchainChecker()
        self.social_analyzer = SocialAnalyzer()
        self.code_analyzer = CodeAnalyzer()
        self.feature_engineer = FeatureEngineer()

        # Modeli yÃ¼kle
        model_path = "models/scam_model.pkl"
        if os.path.exists(model_path):
            self.model = joblib.load(model_path)
        else:
            print("âš ï¸ Model bulunamadÄ±. Ã–nce modeli eÄŸitmeniz gerekiyor.")
            self.model = None

    def analyze_contract(self, contract_address, token_name=""):
        """Ana analiz fonksiyonu"""
        print(f"ðŸ” Analiz baÅŸlÄ±yor: {contract_address}")

        try:
            # 1. Onchain verilerini al
            print("ðŸ“Š Onchain veriler alÄ±nÄ±yor...")
            onchain_features = self.onchain_checker.get_contract_features(contract_address)

            # 2. Sosyal medya analizi (opsiyonel)
            social_features = None
            if token_name:
                print("ðŸ¦ Sosyal medya analizi yapÄ±lÄ±yor...")
                social_features = self.social_analyzer.analyze_social_signals(token_name, contract_address)

            # 3. Kod analizi
            print("ðŸ’» Kontrat kodu analiz ediliyor...")
            source_code = self.onchain_checker._get_contract_source(contract_address).get("source_code", "")
            code_flags = self.code_analyzer.analyze_contract_code(contract_address, source_code)

            # 4. ML Ã¶zelliklerini oluÅŸtur
            features = self.feature_engineer.build_features(onchain_features, social_features)

            # 5. Risk skorunu hesapla
            risk_score = 50  # Default
            ml_prediction = "Model bulunamadÄ±"

            if self.model is not None:
                try:
                    risk_probability = self.model.predict_proba([features])[0][1]
                    risk_score = round(risk_probability * 100, 2)
                    ml_prediction = "Scam" if risk_probability > 0.7 else "GÃ¼venli" if risk_probability < 0.3 else "Dikkatli ol"
                except Exception as e:
                    print(f"Model tahmin hatasÄ±: {e}")

            # 6. SonuÃ§larÄ± birleÅŸtir
            result = {
                "contract": contract_address,
                "token_name": token_name,
                "risk_score": risk_score,
                "ml_prediction": ml_prediction,
                "red_flags": code_flags,
                "onchain_data": onchain_features,
                "social_data": social_features,
                "analysis_time": datetime.now().strftime("%Y-%m-%d %H:%M:%S"),
                "recommendations": self._generate_recommendations(risk_score, onchain_features, code_flags)
            }

            return result

        except Exception as e:
            print(f"âŒ Analiz hatasÄ±: {e}")
            return {
                "contract": contract_address,
                "risk_score": 100,  # Hata durumunda yÃ¼ksek risk
                "ml_prediction": "Analiz edilemedi",
                "red_flags": [f"âŒ Analiz hatasÄ±: {str(e)}"],
                "error": str(e)
            }

    def _generate_recommendations(self, risk_score, onchain_data, code_flags):
        """Risk skoruna gÃ¶re Ã¶neriler oluÅŸtur"""
        recommendations = []

        if risk_score >= 80:
            recommendations.append("ðŸš¨ YÃœKÂ­SEK RÄ°SK: Bu token'a yatÄ±rÄ±m yapmayÄ±n!")
            recommendations.append("ðŸ’¡ BaÅŸka projeleri araÅŸtÄ±rÄ±n")
        elif risk_score >= 60:
            recommendations.append("âš ï¸ ORTA RÄ°SK: Ã‡ok dikkatli olun")
            recommendations.append("ðŸ’¡ KÃ¼Ã§Ã¼k miktarla test edin")
            recommendations.append("ðŸ’¡ Exit stratejinizi belirleyin")
        elif risk_score >= 40:
            recommendations.append("ðŸŸ¡ DÃœÂ­ÅžÃœK-ORTA RÄ°SK: AraÅŸtÄ±rma yapÄ±n")
            recommendations.append("ðŸ’¡ Topluluk ve geliÅŸtirici aktivitesini kontrol edin")
        else:
            recommendations.append("âœ… DÃœÂ­ÅžÃœK RÄ°SK: GÃ¶rece gÃ¼venli gÃ¶rÃ¼nÃ¼yor")
            recommendations.append("ðŸ’¡ Yine de kendi araÅŸtÄ±rmanÄ±zÄ± yapÄ±n")

        # Ã–zel durumlar iÃ§in Ã¶neriler
        if not onchain_data.get("contract_verified", False):
            recommendations.append("ðŸ’¡ Kontrat kodu doÄŸrulanmamÄ±ÅŸ - ekstra dikkat")

        if onchain_data.get("holder_concentration", 0) > 50:
            recommendations.append("ðŸ’¡ Token daÄŸÄ±lÄ±mÄ± centralized - whale riski")

        if not onchain_data.get("owner_renounced", False):
            recommendations.append("ðŸ’¡ Owner haklarÄ± hala aktif - kontrol riski")

        return recommendations

def main():
    """Test fonksiyonu"""
    detector = ScamDetector()

    # Test adresi (Ã¶rnek)
    test_address = "0xA0b86991c31cB32c05C6f5F1B0a5b4C2a5d4c0a6"  # USDC

    result = detector.analyze_contract(test_address, "USDC")

    print("\n" + "="*50)
    print("ðŸ¤– SCAM BOT ANALYSIS SONUÃ‡LARI")
    print("="*50)
    print(f"ðŸ“§ Kontrat: {result['contract']}")
    print(f"ðŸŽ¯ Risk Skoru: {result['risk_score']}%")
    print(f"ðŸ¤– ML Tahmini: {result['ml_prediction']}")
    print(f"ðŸ“… Analiz ZamanÄ±: {result.get('analysis_time', 'Bilinmiyor')}")

    print(f"\nðŸš© Risk BayraklarÄ±:")
    for flag in result['red_flags']:
        print(f"  {flag}")

    print(f"\nðŸ’¡ Ã–neriler:")
    for rec in result.get('recommendations', []):
        print(f"  {rec}")

    if result.get('onchain_data'):
        print(f"\nðŸ“Š Onchain Veriler:")
        data = result['onchain_data']
        print(f"  â€¢ Holder Konsantrasyonu: %{data.get('holder_concentration', 0):.1f}")
        print(f"  â€¢ Mint Fonksiyonu: {'Var' if data.get('mint_function') else 'Yok'}")
        print(f"  â€¢ LP Kilidi: {'Var' if data.get('lp_locked') else 'Yok'}")
        print(f"  â€¢ Maksimum Tax: %{data.get('max_tax', 0)}")
        print(f"  â€¢ Owner Renounced: {'Evet' if data.get('owner_renounced') else 'HayÄ±r'}")
        print(f"  â€¢ Kontrat DoÄŸrulanmÄ±ÅŸ: {'Evet' if data.get('contract_verified') else 'HayÄ±r'}")

if __name__ == "__main__":
    main()
```

```python
%%writefile utils/onchain_checker.py
import requests
from web3 import Web3
from config import Config
import json
import time

class OnchainChecker:
    def __init__(self):
        self.w3 = Web3(Web3.HTTPProvider(Config.INFURA_URL))
        self.etherscan_api = Config.ETHERSCAN_API_KEY
        # Check Infura connection
        if not self.w3.is_connected():
            print("âŒ Infura baÄŸlantÄ±sÄ± kurulamadÄ±! INFURA_PROJECT_ID'nizi kontrol edin.")
        else:
            print("âœ… Infura baÄŸlantÄ±sÄ± baÅŸarÄ±lÄ±.")


    def get_contract_features(self, address):
        try:
            # Kontrat adresini doÄŸrula
            if not self.w3.is_address(address):
                raise ValueError("GeÃ§ersiz kontrat adresi")

            address = self.w3.to_checksum_address(address)
            print(f"ðŸ” Analiz edilen adres (checksum): {address}")


            # Token bilgilerini al
            print("-> Token bilgileri alÄ±nÄ±yor...")
            token_info = self._get_token_info(address)
            print(f"<- Token bilgileri: {token_info}")

            print("-> Holder daÄŸÄ±lÄ±mÄ± alÄ±nÄ±yor...")
            holder_data = self._get_holder_distribution(address)
            print(f"<- Holder sayÄ±sÄ±: {len(holder_data)}")


            print("-> Kontrat kaynak kodu alÄ±nÄ±yor...")
            contract_code = self._get_contract_source(address)
            print(f"<- Kontrat doÄŸrulanmÄ±ÅŸ: {contract_code.get('verified', False)}")


            print("-> Likidite kilidi kontrol ediliyor...")
            liquidity_info = self._check_liquidity_lock(address)
            print(f"<- LP kilitli: {liquidity_info.get('locked', False)}")


            print("-> Owner renounce kontrol ediliyor...")
            owner_renounced = self._check_ownership_renounced(address)
            print(f"<- Owner renounce edilmiÅŸ: {owner_renounced}")


            print("-> Kontrat oluÅŸturulma tarihi alÄ±nÄ±yor...")
            creation_date = self._get_creation_date(address)
            print(f"<- OluÅŸturulma tarihi (timestamp): {creation_date}")


            return {
                "holder_concentration": self._calculate_concentration(holder_data),
                "mint_function": self._check_mint_function(contract_code),
                "lp_locked": liquidity_info.get("locked", False),
                "max_tax": self._extract_max_tax(contract_code),
                "owner_renounced": owner_renounced,
                "contract_verified": contract_code.get("verified", False),
                "creation_date": creation_date,
                "total_supply": token_info.get("total_supply", 0),
                "holder_count": len(holder_data)
            }
        except Exception as e:
            print(f"âŒ OnchainChecker hatasÄ±: {e}")
            return self._get_default_features()

    def _get_token_info(self, address):
        """Token temel bilgilerini al"""
        try:
            # ERC-20 ABI (sadece gerekli fonksiyonlar)
            erc20_abi = [
                {
                    "constant": True,
                    "inputs": [],
                    "name": "totalSupply",
                    "outputs": [{"name": "", "type": "uint256"}],
                    "type": "function"
                }
            ]

            contract = self.w3.eth.contract(address=address, abi=erc20_abi)
            total_supply = contract.functions.totalSupply().call()

            return {"total_supply": total_supply}
        except Exception as e:
            print(f"  _get_token_info hatasÄ±: {e}")
            return {"total_supply": 0}

    def _get_holder_distribution(self, address):
        """Holder daÄŸÄ±lÄ±mÄ±nÄ± Etherscan'den al"""
        try:
            url = f"https://api.etherscan.io/api"
            params = {
                "module": "token",
                "action": "tokenholderlist",
                "contractaddress": address,
                "page": 1,
                "offset": 100,
                "apikey": self.etherscan_api
            }
            # requests handles urlencoding parameters automatically
            print(f"  Etherscan API Ã§aÄŸrÄ±sÄ± (tokenholderlist): {url}?module=token&action=tokenholderlist&contractaddress={address}&page=1&offset=100&apikey=...")

            response = requests.get(url, params=params, timeout=15) # Timeout artÄ±rÄ±ldÄ±
            response.raise_for_status() # HTTP hatalarÄ±nÄ± yakala
            data = response.json()
            print(f"  Etherscan API yanÄ±tÄ± (tokenholderlist status): {data.get('status')}, message: {data.get('message')}")


            if data.get("status") == "1":
                return data.get("result", [])
            else:
                print(f"  Etherscan API'den holder bilgisi alÄ±namadÄ±: {data.get('message', 'Bilinmeyen hata')}")
                return []
        except requests.exceptions.RequestException as e:
            print(f"  _get_holder_distribution istek hatasÄ±: {e}")
            return []
        except Exception as e:
            print(f"  _get_holder_distribution genel hata: {e}")
            return []


    def _calculate_concentration(self, holders):
        """Top 10 holder konsantrasyonunu hesapla"""
        if not holders:
            return 100  # Veri yoksa risk yÃ¼ksek

        try:
            # Ä°lk 10 holder'Ä±n toplam token yÃ¼zdesi
            # TokenHolderQuantity string gelebilir, float'a Ã§evirirken hata olabilir
            top_10_balance = 0
            for h in holders[:10]:
                try:
                    top_10_balance += float(h.get("TokenHolderQuantity", 0))
                except ValueError:
                    print(f"  _calculate_concentration ValueError: {h.get('TokenHolderQuantity')} float'a Ã§evrilemedi.")
                    continue # HatalÄ± veriyi atla

            total_supply = 0
            for h in holders:
                 try:
                    total_supply += float(h.get("TokenHolderQuantity", 0))
                 except ValueError:
                     print(f"  _calculate_concentration ValueError: {h.get('TokenHolderQuantity')} float'a Ã§evrilemedi.")
                     continue # HatalÄ± veriyi atla


            if total_supply == 0:
                return 100

            concentration = (top_10_balance / total_supply) * 100
            return min(concentration, 100)  # Max %100
        except Exception as e:
            print(f"  _calculate_concentration genel hata: {e}")
            return 75  # Default deÄŸer

    def _get_contract_source(self, address):
        """Kontrat kaynak kodunu al"""
        try:
            url = f"https://api.etherscan.io/api"
            params = {
                "module": "contract",
                "action": "getsourcecode",
                "address": address,
                "apikey": self.etherscan_api
            }
            # requests handles urlencoding parameters automatically
            print(f"  Etherscan API Ã§aÄŸrÄ±sÄ± (getsourcecode): {url}?module=contract&action=getsourcecode&address={address}&apikey=...")


            response = requests.get(url, params=params, timeout=15) # Timeout artÄ±rÄ±ldÄ±
            response.raise_for_status() # HTTP hatalarÄ±nÄ± yakala
            data = response.json()
            print(f"  Etherscan API yanÄ±tÄ± (getsourcecode status): {data.get('status')}, message: {data.get('message')}")


            if data.get("status") == "1" and data.get("result"):
                source_code = data["result"][0].get("SourceCode", "")
                return {
                    "source_code": source_code,
                    "verified": len(source_code) > 0 and source_code != "Contract source code not verified" # DoÄŸrulama mesajÄ±nÄ± da kontrol et
                }
            else:
                 print(f"  Etherscan API'den kaynak kodu alÄ±namadÄ±: {data.get('message', 'Bilinmeyen hata')}")
                 return {"source_code": "", "verified": False}

        except requests.exceptions.RequestException as e:
            print(f"  _get_contract_source istek hatasÄ±: {e}")
            return {"source_code": "", "verified": False}
        except Exception as e:
            print(f"  _get_contract_source genel hata: {e}")
            return {"source_code": "", "verified": False}


    def _check_mint_function(self, contract_data):
        """Mint fonksiyonu var mÄ± kontrol et"""
        source_code = contract_data.get("source_code", "").lower()
        mint_keywords = ["function mint", "mint(", "_mint(", "mintto", "mint_"]
        return any(keyword in source_code for keyword in mint_keywords)

    def _extract_max_tax(self, contract_data):
        """Maksimum tax oranÄ±nÄ± bulmaya Ã§alÄ±ÅŸ"""
        source_code = contract_data.get("source_code", "").lower()

        # YaygÄ±n tax variable isimleri
        tax_patterns = ["buytax", "selltax", "tax", "fee"]

        import re
        max_tax = 0
        for pattern in tax_patterns:
            # uint256 public buyTax = 5; gibi patternleri ara
            regex = rf"{pattern}\s*=\s*(\d+)"
            matches = re.findall(regex, source_code)
            for match in matches:
                 try:
                     max_tax = max(max_tax, int(match))
                 except ValueError:
                     print(f"  _extract_max_tax ValueError: {match} int'e Ã§evrilemedi.")
                     continue # HatalÄ± veriyi atla

        return max_tax # Default deÄŸer


    def _check_ownership_renounced(self, address):
        """Owner renounce edilmiÅŸ mi kontrol et"""
        try:
            # Owner fonksiyonu Ã§aÄŸÄ±r
            owner_abi = [{
                "constant": True,
                "inputs": [],
                "name": "owner",
                "outputs": [{"name": "", "type": "address"}],
                "type": "function"
            }]

            contract = self.w3.eth.contract(address=address, abi=owner_abi)
            owner = contract.functions.owner().call()

            # 0x000...000 adresine sahipse renounce edilmiÅŸ
            return owner == "0x0000000000000000000000000000000000000000"
        except Exception as e:
            print(f"  _check_ownership_renounced hatasÄ±: {e}")
            return False

    def _check_liquidity_lock(self, address):
        """Likidite kilidi kontrol et (basit versiyon)"""
        # Bu gerÃ§ek implementasyonda Uniswap/PancakeSwap pool kontratlarÄ±nÄ± kontrol etmeli
        # Etherscan API'sinde doÄŸrudan likidite kilidi bilgisi yok, bu daha karmaÅŸÄ±k bir analiz gerektirir.
        # Åžimdilik sabit deÄŸer dÃ¶ndÃ¼rÃ¼yoruz.
        print("  _check_liquidity_lock: Basit kontrol, her zaman False dÃ¶ndÃ¼rÃ¼yor.")
        return {"locked": False, "lock_time": 0}

    def _get_creation_date(self, address):
        """Kontrat oluÅŸturulma tarihini al"""
        try:
            url = f"https://api.etherscan.io/api"
            params = {
                "module": "account",
                "action": "txlist",
                "address": address,
                "startblock": 0,
                "endblock": 99999999,
                "page": 1,
                "offset": 1,
                "sort": "asc",
                "apikey": self.etherscan_api
            }
            # requests handles urlencoding parameters automatically
            print(f"  Etherscan API Ã§aÄŸrÄ±sÄ± (txlist): {url}?module=account&action=txlist&address={address}&startblock=0&endblock=99999999&page=1&offset=1&sort=asc&apikey=...")


            response = requests.get(url, params=params, timeout=15) # Timeout artÄ±rÄ±ldÄ±
            response.raise_for_status() # HTTP hatalarÄ±nÄ± yakala
            data = response.json()
            print(f"  Etherscan API yanÄ±tÄ± (txlist status): {data.get('status')}, message: {data.get('message')}")


            if data.get("status") == "1" and data.get("result"):
                timestamp = int(data["result"][0].get("timeStamp", 0))
                return timestamp
            else:
                print(f"  Etherscan API'den oluÅŸturulma tarihi alÄ±namadÄ±: {data.get('message', 'Bilinmeyen hata')}")
                return 0
        except requests.exceptions.RequestException as e:
            print(f"  _get_creation_date istek hatasÄ±: {e}")
            return 0
        except Exception as e:
            print(f"  _get_creation_date genel hata: {e}")
            return 0


    def _get_default_features(self):
        """Hata durumunda default deÄŸerler"""
        return {
            "holder_concentration": 75,
            "mint_function": True,
            "lp_locked": False,
            "max_tax": 25,
            "owner_renounced": False,
            "contract_verified": False,
            "creation_date": 0,
            "total_supply": 0,
            "holder_count": 0
        }
```

```python
%%writefile utils/social_analysis.py
import requests
import tweepy
from config import Config
from datetime import datetime, timedelta

class SocialAnalyzer:
    def __init__(self):
        if Config.TWITTER_BEARER_TOKEN:
            try:
                self.twitter_client = tweepy.Client(bearer_token=Config.TWITTER_BEARER_TOKEN)
                # Verify credentials - basic check
                # self.twitter_client.get_me() # Bu kotaya dahil olabilir, dikkatli kullanÄ±n
                print("âœ… Twitter istemcisi baÅŸlatÄ±ldÄ±.")
            except Exception as e:
                 print(f"âŒ Twitter istemcisi baÅŸlatÄ±lÄ±rken hata oluÅŸtu: {e}")
                 self.twitter_client = None
        else:
            print("âš ï¸ TWITTER_BEARER_TOKEN ayarlanmamÄ±ÅŸ. Sosyal medya analizi yapÄ±lamayacak.")
            self.twitter_client = None


    def analyze_social_signals(self, token_name, contract_address):
        """Token iÃ§in sosyal medya sinyallerini analiz et"""
        signals = {
            "twitter_mentions": 0,
            "sentiment_score": 0.5,  # 0-1 arasÄ±
            "bot_activity": False,
            "pump_dump_keywords": False
        }

        if self.twitter_client:
            try:
                # Sorgu oluÅŸturma - Hata ayÄ±klama iÃ§in basÄ±ldÄ±
                # Tweet sorgusu iÃ§in kurallar: https://developer.twitter.com/en/docs/twitter-api/tweets/search/integrate/build-a-query
                # Contract adresi yerine sadece token adÄ± veya kÄ±saltmasÄ± daha iyi olabilir
                # Ã‡ok kÄ±sa contract adresi parÃ§asÄ± ([:10]) geÃ§ersiz sorguya neden olabilir
                # Token adÄ± boÅŸsa veya Ã§ok kÄ±saysa da sorun olabilir.
                query_parts = []
                # Check if token_name is provided and is reasonably long
                if token_name and len(token_name.strip()) > 2: # Use token name if provided and > 2 chars
                    query_parts.append(f'"{token_name.strip()}"')
                # Add contract address only if token_name is not used or is short, and address is valid length
                # Using the full address in query might be too specific, maybe just the beginning?
                # Or perhaps only use contract address if no token name is given?
                # Let's prioritize token name, then fall back to part of address if name is not useful.
                elif contract_address and len(contract_address) > 10: # Check contract address length
                     query_parts.append(f'{contract_address[:10]}') # Use first 10 chars


                if not query_parts:
                     print("âš ï¸ Twitter sorgusu iÃ§in yeterli bilgi (token adÄ± veya kontrat adresi) yok.")
                     return signals # BoÅŸ sinyaller dÃ¶ndÃ¼r

                # Combine query parts with OR, ensure no leading/trailing OR if only one part
                query = " OR ".join(query_parts)
                final_query = f"{query} -is:retweet lang:en" # Retweetleri hariÃ§ tut

                print(f"  Twitter API sorgusu: {final_query}")

                # Son 24 saat iÃ§inde token hakkÄ±nda tweetleri ara
                # start_time = datetime.utcnow() - timedelta(hours=24)
                # start_time_str = start_time.isoformat("T") + "Z" # RFC3339 formatÄ±


                tweets = self.twitter_client.search_recent_tweets(
                    query=final_query,
                    max_results=10, # Test iÃ§in daha az sonuÃ§ Ã§ek
                    # start_time=start_time_str, # Son 24 saat
                    tweet_fields=["created_at", "author_id", "public_metrics", "text"] # text'i de al
                )

                if tweets and tweets.data:
                    print(f"  Twitter API {len(tweets.data)} tweet buldu.")
                    signals["twitter_mentions"] = len(tweets.data)
                    signals["pump_dump_keywords"] = self._check_pump_dump_keywords(tweets.data)
                    signals["bot_activity"] = self._detect_bot_activity(tweets.data)
                    signals["sentiment_score"] = self._calculate_sentiment(tweets.data)
                else:
                    print("  Twitter API tweet bulamadÄ±.")
                    signals["twitter_mentions"] = 0


            except tweepy.errors.TweepyException as e:
                 print(f"âŒ Twitter API hatasÄ± (TweepyException): {e}")
                 # Hata detaylarÄ±nÄ± yazdÄ±rma
                 if hasattr(e, 'response') and e.response is not None:
                     print(f"  HTTP Status Code: {e.response.status_code}")
                     try:
                         print(f"  Response Body: {e.response.json()}")
                     except:
                         print(f"  Response Body: {e.response.text}")

            except Exception as e:
                print(f"âŒ Sosyal medya analizi genel hata: {e}")

        else:
            print("âš ï¸ Twitter istemcisi kullanÄ±lamÄ±yor. Sosyal medya analizi atlandÄ±.")


        return signals

    def _check_pump_dump_keywords(self, tweets):
        """Pump&dump anahtar kelimelerini kontrol et"""
        pump_keywords = [
            "moon", "ðŸš€", "pump", "easy money", "quick profit",
            "100x", "1000x", "get rich", "diamond hands", "ape in"
        ]

        # tweets bir liste, her Ã¶ÄŸe bir Tweet nesnesi
        tweet_texts = " ".join([tweet.text.lower() for tweet in tweets])
        return any(keyword in tweet_texts for keyword in pump_keywords)

    def _detect_bot_activity(self, tweets):
        """Bot aktivitesi tespit et"""
        if len(tweets) < 5: # Bot aktivitesi iÃ§in minimum tweet sayÄ±sÄ±
            return False

        # AynÄ± metinlerin tekrarÄ±
        tweet_texts = [tweet.text.strip() for tweet in tweets] # BaÅŸÄ±ndaki ve sonundaki boÅŸluklarÄ± kaldÄ±r
        unique_tweets = set(tweet_texts)

        # %50'den fazlasÄ± aynÄ±ysa bot aktivitesi olabilir (daha dÃ¼ÅŸÃ¼k eÅŸik)
        is_bot = len(unique_tweets) / len(tweet_texts) < 0.5
        if is_bot:
            print("  Bot aktivitesi ÅŸÃ¼phesi: Tekrarlayan tweetler tespit edildi.")
        return is_bot


    def _calculate_sentiment(self, tweets):
        """Basit sentiment analizi"""
        positive_words = ["good", "great", "amazing", "bullish", "buy", "long", "holding", "hodl", "strong"]
        negative_words = ["scam", "rug", "fake", "avoid", "dump", "short", "sell", "warning", "risk"]

        total_score = 0
        if not tweets:
            return 0.5 # Tweet yoksa nÃ¶tr

        for tweet in tweets:
            text = tweet.text.lower()
            pos_count = sum(word in text for word in positive_words)
            neg_count = sum(word in text for word in negative_words)

            # Sentiment skorunu hesapla (basit oran)
            score = 0.5 # Default nÃ¶tr
            if pos_count > neg_count:
                score = 1.0 # Pozitif
            elif neg_count > pos_count:
                score = 0.0 # Negatif
            # EÄŸer pos_count == neg_count ise skor 0.5 kalÄ±r (nÃ¶tr)

            total_score += score

        # Ortalama sentiment
        average_sentiment = total_score / len(tweets)
        return average_sentiment
```

```python
%%writefile utils/code_analysis.py
import re

class CodeAnalyzer:
    def __init__(self):
        self.risk_patterns = {
            "mint_function": [
                r"function\s+mint\s*\(",
                r"function\s+_mint\s*\(",
                r"\.mint\s*\(",
                r"mintTo\s*\("
            ],
            "ownership_issues": [
                r"onlyOwner",
                r"_owner\s*=",
                r"transferOwnership",
                r"renounceOwnership"
            ],
            "liquidity_removal": [
                r"removeLiquidity",
                r"withdraw.*Liquidity",
                r"emergencyWithdraw"
            ],
            "tax_functions": [
                r"buyTax\s*=",
                r"sellTax\s*=",
                r"setTax",
                r"updateTax"
            ],
            "blacklist_functions": [
                r"blacklist",
                r"addBot",
                r"removeBot",
                r"setBot"
            ],
            "pause_functions": [
                r"pause\s*\(",
                r"unpause\s*\(",
                r"setPaused"
            ]
        }

    def analyze_contract_code(self, address, source_code=""):
        """Kontrat kodunu analiz ederek risk bayraklarÄ±nÄ± dÃ¶ndÃ¼r"""
        flags = []

        if not source_code:
            flags.append("âš ï¸ Kontrat kodu doÄŸrulanamadÄ±")
            return flags

        source_code = source_code.lower()

        # Mint fonksiyonu kontrolÃ¼
        if self._check_pattern_exists(source_code, self.risk_patterns["mint_function"]):
            flags.append("ðŸ”´ Mint fonksiyonu tespit edildi")

        # Ownership kontrolleri
        if self._check_pattern_exists(source_code, self.risk_patterns["ownership_issues"]):
            if not self._check_ownership_renounced(source_code):
                flags.append("ðŸ”´ Owner haklarÄ± korunuyor")

        # Likidite Ã§ekme fonksiyonlarÄ±
        if self._check_pattern_exists(source_code, self.risk_patterns["liquidity_removal"]):
            flags.append("ðŸ”´ Likidite Ã§ekme fonksiyonu var")

        # Vergi fonksiyonlarÄ±
        if self._check_pattern_exists(source_code, self.risk_patterns["tax_functions"]):
            tax_rate = self._extract_tax_rate(source_code)
            if tax_rate > 10:
                flags.append(f"ðŸ”´ YÃ¼ksek vergi oranÄ±: %{tax_rate}")

        # Blacklist fonksiyonlarÄ±
        if self._check_pattern_exists(source_code, self.risk_patterns["blacklist_functions"]):
            flags.append("ðŸ”´ Blacklist fonksiyonu tespit edildi")

        # Pause fonksiyonlarÄ±
        if self._check_pattern_exists(source_code, self.risk_patterns["pause_functions"]):
            flags.append("ðŸ”´ Trading durdurma fonksiyonu var")

        # Honeypot kontrolleri
        if self._check_honeypot_patterns(source_code):
            flags.append("ðŸ”´ Honeypot kalÄ±plarÄ± tespit edildi")

        if not flags:
            flags.append("âœ… Kod analizi temiz")

        return flags

    def _check_pattern_exists(self, code, patterns):
        """Verilen pattern'lerden herhangi biri var mÄ± kontrol et"""
        for pattern in patterns:
            if re.search(pattern, code, re.IGNORECASE):
                return True
        return False

    def _check_ownership_renounced(self, code):
        """Ownership renounce edilmiÅŸ mi kontrol et"""
        renounce_patterns = [
            r"renounceownership\s*\(\s*\)",
            r"_owner\s*=.*0x0+",
            r"owner.*=.*address\(0\)"
        ]
        return self._check_pattern_exists(code, renounce_patterns)

    def _extract_tax_rate(self, code):
        """Kod iÃ§inden vergi oranÄ±nÄ± Ã§Ä±karma"""
        tax_patterns = [
            r"buytax\s*=\s*(\d+)",
            r"selltax\s*=\s*(\d+)",
            r"tax.*=\s*(\d+)"
        ]

        max_tax = 0
        for pattern in tax_patterns:
            matches = re.findall(pattern, code, re.IGNORECASE)
            for match in matches:
                max_tax = max(max_tax, int(match))

        return max_tax

    def _check_honeypot_patterns(self, code):
        """Honeypot kalÄ±plarÄ±nÄ± kontrol et"""
        honeypot_patterns = [
            r"require.*balances\[.*\]\s*>=",  # Balance kontrolÃ¼
            r"if.*balances\[.*\].*return false",  # Transfer engelleme
            r"mapping.*bool.*isexcluded",  # Exclusion mapping
            r"tradingopen\s*=\s*false"  # Trading kapalÄ±
        ]
        return self._check_pattern_exists(code, honeypot_patterns)
```

```python
%%writefile utils/feature_engineering.py
import numpy as np
from datetime import datetime

class FeatureEngineer:
    def build_features(self, onchain_data, social_data=None):
        """Ham verilerden ML modeli iÃ§in Ã¶zellikler oluÅŸtur"""

        # Temel Ã¶zellikler
        features = [
            onchain_data.get("holder_concentration", 75),
            int(onchain_data.get("mint_function", True)),
            int(onchain_data.get("lp_locked", False)),
            onchain_data.get("max_tax", 25),
            int(onchain_data.get("owner_renounced", False)),
            int(onchain_data.get("contract_verified", False))
        ]

        # Kontrat yaÅŸÄ± (gÃ¼n cinsinden)
        creation_date = onchain_data.get("creation_date", 0)
        if creation_date > 0:
            age_days = (datetime.now().timestamp() - creation_date) / 86400
            features.append(min(age_days, 365))  # Max 1 yÄ±l
        else:
            features.append(0)

        # Holder sayÄ±sÄ± (log Ã¶lÃ§eÄŸi)
        holder_count = onchain_data.get("holder_count", 1)
        features.append(np.log10(max(holder_count, 1)))

        # Sosyal medya Ã¶zellikleri
        if social_data:
            features.extend([
                social_data.get("twitter_mentions", 0),
                social_data.get("sentiment_score", 0.5) * 100,  # 0-100 Ã¶lÃ§eÄŸi
                int(social_data.get("bot_activity", False)),
                int(social_data.get("pump_dump_keywords", False))
            ])
        else:
            features.extend([0, 50, 0, 0])  # Default deÄŸerler

        return np.array(features)

    def get_feature_names(self):
        """Ã–zellik isimlerini dÃ¶ndÃ¼r"""
        return [
            "holder_concentration",
            "mint_function",
            "lp_locked",
            "max_tax",
            "owner_renounced",
            "contract_verified",
            "age_days",
            "log_holder_count",
            "twitter_mentions",
            "sentiment_score",
            "bot_activity",
            "pump_dump_keywords"
        ]
```

```python
%%writefile main.py

import gradio as gr
import sys
import os
import requests
from web3 import Web3
from dotenv import load_dotenv
import json
import time
import numpy as np
import pandas as pd
from datetime import datetime
import re
import tweepy # Sadece Config.TWITTER_BEARER_TOKEN varsa kullanÄ±lacak

# Load environment variables from .env
load_dotenv()

class Config:
    INFURA_PROJECT_ID = os.getenv('INFURA_PROJECT_ID')
    ETHERSCAN_API_KEY = os.getenv('ETHERSCAN_API_KEY')
    TWITTER_BEARER_TOKEN = os.getenv('TWITTER_BEARER_TOKEN')
    INFURA_URL = f"https://mainnet.infura.io/v3/{INFURA_PROJECT_ID}"

# utils/onchain_checker.py iÃ§eriÄŸi
class OnchainChecker:
    def __init__(self):
        # API anahtarlarÄ±nÄ±n ayarlandÄ±ÄŸÄ±ndan emin olun
        if not Config.INFURA_PROJECT_ID or not Config.ETHERSCAN_API_KEY:
             print("âŒ API anahtarlarÄ± (Infura, Etherscan) ayarlanmamÄ±ÅŸ! On-chain analiz yapÄ±lamayacak.")
             self.w3 = None # Dummy Web3 objesi
             self.etherscan_api = None
        else:
            self.w3 = Web3(Web3.HTTPProvider(Config.INFURA_URL))
            self.etherscan_api = Config.ETHERSCAN_API_KEY
            # Check Infura connection
            if not self.w3.is_connected():
                print("âŒ Infura baÄŸlantÄ±sÄ± kurulamadÄ±! INFURA_PROJECT_ID'nizi kontrol edin.")
            else:
                print("âœ… Infura baÄŸlantÄ±sÄ± baÅŸarÄ±lÄ±.")


    def get_contract_features(self, address):
        if self.w3 is None or self.etherscan_api is None:
            print("âŒ API anahtarlarÄ± eksik, on-chain analiz atlandÄ±.")
            return self._get_default_features()


        try:
            # Kontrat adresini doÄŸrula
            if not self.w3.is_address(address):
                raise ValueError("GeÃ§ersiz kontrat adresi")

            address = self.w3.to_checksum_address(address)
            print(f"ðŸ” Analiz edilen adres (checksum): {address}")


            # Token bilgilerini al
            print("-> Token bilgileri alÄ±nÄ±yor...")
            token_info = self._get_token_info(address)
            print(f"<- Token bilgileri: {token_info}")

            print("-> Holder daÄŸÄ±lÄ±mÄ± alÄ±nÄ±yor...")
            holder_data = self._get_holder_distribution(address)
            print(f"<- Holder sayÄ±sÄ±: {len(holder_data)}")


            print("-> Kontrat kaynak kodu alÄ±nÄ±yor...")
            contract_code = self._get_contract_source(address)
            print(f"<- Kontrat doÄŸrulanmÄ±ÅŸ: {contract_code.get('verified', False)}")


            print("-> Likidite kilidi kontrol ediliyor...")
            liquidity_info = self._check_liquidity_lock(address)
            print(f"<- LP kilitli: {liquidity_info.get('locked', False)}")


            print("-> Owner renounce kontrol ediliyor...")
            owner_renounced = self._check_ownership_renounced(address)
            print(f"<- Owner renounce edilmiÅŸ: {owner_renounced}")


            print("-> Kontrat oluÅŸturulma tarihi alÄ±nÄ±yor...")
            creation_date = self._get_creation_date(address)
            print(f"<- OluÅŸturulma tarihi (timestamp): {creation_date}")


            return {
                "holder_concentration": self._calculate_concentration(holder_data),
                "mint_function": self._check_mint_function(contract_code),
                "lp_locked": liquidity_info.get("locked", False),
                "max_tax": self._extract_max_tax(contract_code),
                "owner_renounced": owner_renounced,
                "contract_verified": contract_code.get("verified", False),
                "creation_date": creation_date,
                "total_supply": token_info.get("total_supply", 0),
                "holder_count": len(holder_data)
            }
        except Exception as e:
            print(f"âŒ OnchainChecker hatasÄ±: {e}")
            return self._get_default_features()

    def _get_token_info(self, address):
        """Token temel bilgilerini al"""
        if self.w3 is None: return {"total_supply": 0}
        try:
            # ERC-20 ABI (sadece gerekli fonksiyonlar)
            erc20_abi = [
                {
                    "constant": True,
                    "inputs": [],
                    "name": "totalSupply",
                    "outputs": [{"name": "", "type": "uint256"}],
                    "type": "function"
                }
            ]

            contract = self.w3.eth.contract(address=address, abi=erc20_abi)
            total_supply = contract.functions.totalSupply().call()

            return {"total_supply": total_supply}
        except Exception as e:
            print(f"  _get_token_info hatasÄ±: {e}")
            return {"total_supply": 0}

    def _get_holder_distribution(self, address):
        """Holder daÄŸÄ±lÄ±mÄ±nÄ± Etherscan'den al"""
        if self.etherscan_api is None: return []
        try:
            url = f"https://api.etherscan.io/api"
            params = {
                "module": "token",
                "action": "tokenholderlist",
                "contractaddress": address,
                "page": 1,
                "offset": 100,
                "apikey": self.etherscan_api
            }
            # requests handles urlencoding parameters automatically
            # print(f"  Etherscan API Ã§aÄŸrÄ±sÄ± (tokenholderlist): {url}?module=token&action=tokenholderlist&contractaddress={address}&page=1&offset=100&apikey=...")

            response = requests.get(url, params=params, timeout=15) # Timeout artÄ±rÄ±ldÄ±
            response.raise_for_status() # HTTP hatalarÄ±nÄ± yakala
            data = response.json()
            # print(f"  Etherscan API yanÄ±tÄ± (tokenholderlist status): {data.get('status')}, message: {data.get('message')}")


            if data.get("status") == "1":
                return data.get("result", [])
            else:
                print(f"  Etherscan API'den holder bilgisi alÄ±namadÄ±: {data.get('message', 'Bilinmeyen hata')}")
                return []
        except requests.exceptions.RequestException as e:
            print(f"  _get_holder_distribution istek hatasÄ±: {e}")
            return []
        except Exception as e:
            print(f"  _get_holder_distribution genel hata: {e}")
            return []


    def _calculate_concentration(self, holders):
        """Top 10 holder konsantrasyonunu hesapla"""
        if not holders:
            return 100  # Veri yoksa risk yÃ¼ksek

        try:
            # Ä°lk 10 holder'Ä±n toplam token yÃ¼zdesi
            # TokenHolderQuantity string gelebilir, float'a Ã§evirirken hata olabilir
            top_10_balance = 0
            for h in holders[:10]:
                try:
                    top_10_balance += float(h.get("TokenHolderQuantity", 0))
                except ValueError:
                    print(f"  _calculate_concentration ValueError: {h.get('TokenHolderQuantity')} float'a Ã§evrilemedi.")
                    continue # HatalÄ± veriyi atla

            total_supply = 0
            for h in holders:
                 try:
                    total_supply += float(h.get("TokenHolderQuantity", 0))
                 except ValueError:
                     print(f"  _calculate_concentration ValueError: {h.get('TokenHolderQuantity')} float'a Ã§evrilemedi.")
                     continue # HatalÄ± veriyi atla


            if total_supply == 0:
                return 100

            concentration = (top_10_balance / total_supply) * 100
            return min(concentration, 100)  # Max %100
        except Exception as e:
            print(f"  _calculate_concentration genel hata: {e}")
            return 75  # Default deÄŸer

    def _get_contract_source(self, address):
        """Kontrat kaynak kodunu al"""
        if self.etherscan_api is None: return {"source_code": "", "verified": False}
        try:
            url = f"https://api.etherscan.io/api"
            params = {
                "module": "contract",
                "action": "getsourcecode",
                "address": address,
                "apikey": self.etherscan_api
            }
            # requests handles urlencoding parameters automatically
            # print(f"  Etherscan API Ã§aÄŸrÄ±sÄ± (getsourcecode): {url}?module=contract&action=getsourcecode&address={address}&apikey=...")


            response = requests.get(url, params=params, timeout=15) # Timeout artÄ±rÄ±ldÄ±
            response.raise_for_status() # HTTP hatalarÄ±nÄ± yakala
            data = response.json()
            # print(f"  Etherscan API yanÄ±tÄ± (getsourcecode status): {data.get('status')}, message: {data.get('message')}")


            if data.get("status") == "1" and data.get("result"):
                source_code = data["result"][0].get("SourceCode", "")
                return {
                    "source_code": source_code,
                    "verified": len(source_code) > 0 and source_code != "Contract source code not verified" # DoÄŸrulama mesajÄ±nÄ± da kontrol et
                }
            else:
                 print(f"  Etherscan API'den kaynak kodu alÄ±namadÄ±: {data.get('message', 'Bilinmeyen hata')}")
                 return {"source_code": "", "verified": False}

        except requests.exceptions.RequestException as e:
            print(f"  _get_contract_source istek hatasÄ±: {e}")
            return {"source_code": "", "verified": False}
        except Exception as e:
            print(f"  _get_contract_source genel hata: {e}")
            return {"source_code": "", "verified": False}


    def _check_mint_function(self, contract_data):
        """Mint fonksiyonu var mÄ± kontrol et"""
        source_code = contract_data.get("source_code", "").lower()
        mint_keywords = ["function mint", "mint(", "_mint(", "mintto", "mint_"]
        return any(keyword in source_code for keyword in mint_keywords)

    def _extract_max_tax(self, contract_data):
        """Maksimum tax oranÄ±nÄ± bulmaya Ã§alÄ±ÅŸ"""
        source_code = contract_data.get("source_code", "").lower()

        # YaygÄ±n tax variable isimleri
        tax_patterns = ["buytax", "selltax", "tax", "fee"]

        import re
        max_tax = 0
        for pattern in tax_patterns:
            # uint256 public buyTax = 5; gibi patternleri ara
            regex = rf"{pattern}\s*=\s*(\d+)"
            matches = re.findall(regex, source_code)
            for match in matches:
                 try:
                     max_tax = max(max_tax, int(match))
                 except ValueError:
                     print(f"  _extract_max_tax ValueError: {match} int'e Ã§evrilemedi.")
                     continue # HatalÄ± veriyi atla

        return max_tax # Default deÄŸer


    def _check_ownership_renounced(self, address):
        """Owner renounce edilmiÅŸ mi kontrol et"""
        if self.w3 is None: return False
        try:
            # Owner fonksiyonu Ã§aÄŸÄ±r
            owner_abi = [{
                "constant": True,
                "inputs": [],
                "name": "owner",
                "outputs": [{"name": "", "type": "address"}],
                "type": "function"
            }]

            contract = self.w3.eth.contract(address=address, abi=owner_abi)
            owner = contract.functions.owner().call()

            # 0x000...000 adresine sahipse renounce edilmiÅŸ
            return owner == "0x0000000000000000000000000000000000000000"
        except Exception as e:
            print(f"  _check_ownership_renounced hatasÄ±: {e}")
            return False

    def _check_liquidity_lock(self, address):
        """Likidite kilidi kontrol et (basit versiyon)"""
        # Bu gerÃ§ek implementasyonda Uniswap/PancakeSwap pool kontratlarÄ±nÄ± kontrol etmeli
        # Etherscan API'sinde doÄŸrudan likidite kilidi bilgisi yok, bu daha karmaÅŸÄ±k bir analiz gerektirir.
        # Åžimdilik sabit deÄŸer dÃ¶ndÃ¼rÃ¼yoruz.
        # print("  _check_liquidity_lock: Basit kontrol, her zaman False dÃ¶ndÃ¼rÃ¼yor.")
        return {"locked": False, "lock_time": 0}

    def _get_creation_date(self, address):
        """Kontrat oluÅŸturulma tarihini al"""
        if self.etherscan_api is None: return 0
        try:
            url = f"https://api.etherscan.io/api"
            params = {
                "module": "account",
                "action": "txlist",
                "address": address,
                "startblock": 0,
                "endblock": 99999999,
                "page": 1,
                "offset": 1,
                "sort": "asc",
                "apikey": self.etherscan_api
            }
            # requests handles urlencoding parameters automatically
            # print(f"  Etherscan API Ã§aÄŸrÄ±sÄ± (txlist): {url}?module=account&action=txlist&address={address}&startblock=0&endblock=99999999&page=1&offset=1&sort=asc&apikey=...")


            response = requests.get(url, params=params, timeout=15) # Timeout artÄ±rÄ±ldÄ±
            response.raise_for_status() # HTTP hatalarÄ±nÄ± yakala
            data = response.json()
            # print(f"  Etherscan API yanÄ±tÄ± (txlist status): {data.get('status')}, message: {data.get('message')}")


            if data.get("status") == "1" and data.get("result"):
                timestamp = int(data["result"][0].get("timeStamp", 0))
                return timestamp
            else:
                print(f"  Etherscan API'den oluÅŸturulma tarihi alÄ±namadÄ±: {data.get('message', 'Bilinmeyen hata')}")
                return 0
        except requests.exceptions.RequestException as e:
            print(f"  _get_creation_date istek hatasÄ±: {e}")
            return 0
        except Exception as e:
            print(f"  _get_creation_date genel hata: {e}")
            return 0


    def _get_default_features(self):
        """Hata durumunda default deÄŸerler"""
        return {
            "holder_concentration": 75,
            "mint_function": True,
            "lp_locked": False,
            "max_tax": 25,
            "owner_renounced": False,
            "contract_verified": False,
            "creation_date": 0,
            "total_supply": 0,
            "holder_count": 0
        }

# utils/social_analysis.py iÃ§eriÄŸi
class SocialAnalyzer:
    def __init__(self):
        if Config.TWITTER_BEARER_TOKEN:
            try:
                self.twitter_client = tweepy.Client(bearer_token=Config.TWITTER_BEARER_TOKEN)
                # Verify credentials - basic check
                # self.twitter_client.get_me() # Bu kotaya dahil olabilir, dikkatli kullanÄ±n
                print("âœ… Twitter istemcisi baÅŸlatÄ±ldÄ±.")
            except Exception as e:
                 print(f"âŒ Twitter istemcisi baÅŸlatÄ±lÄ±rken hata oluÅŸtu: {e}")
                 self.twitter_client = None
        else:
            print("âš ï¸ TWITTER_BEARER_TOKEN ayarlanmamÄ±ÅŸ. Sosyal medya analizi yapÄ±lamayacak.")
            self.twitter_client = None


    def analyze_social_signals(self, token_name, contract_address):
        """Token iÃ§in sosyal medya sinyallerini analiz et"""
        signals = {
            "twitter_mentions": 0,
            "sentiment_score": 0.5,  # 0-1 arasÄ±
            "bot_activity": False,
            "pump_dump_keywords": False
        }

        if self.twitter_client:
            try:
                # Sorgu oluÅŸturma - Hata ayÄ±klama iÃ§in basÄ±ldÄ±
                # Tweet sorgusu iÃ§in kurallar: https://developer.twitter.com/en/docs/twitter-api/tweets/search/integrate/build-a-query
                # Contract adresi yerine sadece token adÄ± veya kÄ±saltmasÄ± daha iyi olabilir
                # Ã‡ok kÄ±sa contract adresi parÃ§asÄ± ([:10]) geÃ§ersiz sorguya neden olabilir
                # Token adÄ± boÅŸsa veya Ã§ok kÄ±saysa da sorun olabilir.
                query_parts = []
                # Check if token_name is provided and is reasonably long
                if token_name and len(token_name.strip()) > 2: # Use token name if provided and > 2 chars
                    query_parts.append(f'"{token_name.strip()}"')
                # Add contract address only if token_name is not used or is short, and address is valid length
                # Using the full address in query might be too specific, maybe just the beginning?
                # Or perhaps only use contract address if no token name is given?
                # Let's prioritize token name, then fall back to part of address if name is not useful.
                elif contract_address and len(contract_address) > 10: # Check contract address length
                     query_parts.append(f'{contract_address[:10]}') # Use first 10 chars


                if not query_parts:
                     print("âš ï¸ Twitter sorgusu iÃ§in yeterli bilgi (token adÄ± veya kontrat adresi) yok.")
                     return signals # BoÅŸ sinyaller dÃ¶ndÃ¼r

                # Combine query parts with OR, ensure no leading/trailing OR if only one part
                query = " OR ".join(query_parts)
                final_query = f"{query} -is:retweet lang:en" # Retweetleri hariÃ§ tut

                print(f"  Twitter API sorgusu: {final_query}")

                # Son 24 saat iÃ§inde token hakkÄ±nda tweetleri ara
                # start_time = datetime.utcnow() - timedelta(hours=24)
                # start_time_str = start_time.isoformat("T") + "Z" # RFC3339 formatÄ±


                tweets = self.twitter_client.search_recent_tweets(
                    query=final_query,
                    max_results=10, # Test iÃ§in daha az sonuÃ§ Ã§ek
                    # start_time=start_time_str, # Son 24 saat
                    tweet_fields=["created_at", "author_id", "public_metrics", "text"] # text'i de al
                )

                if tweets and tweets.data:
                    print(f"  Twitter API {len(tweets.data)} tweet buldu.")
                    signals["twitter_mentions"] = len(tweets.data)
                    signals["pump_dump_keywords"] = self._check_pump_dump_keywords(tweets.data)
                    signals["bot_activity"] = self._detect_bot_activity(tweets.data)
                    signals["sentiment_score"] = self._calculate_sentiment(tweets.data)
                else:
                    print("  Twitter API tweet bulamadÄ±.")
                    signals["twitter_mentions"] = 0


            except tweepy.errors.TweepyException as e:
                 print(f"âŒ Twitter API hatasÄ± (TweepyException): {e}")
                 # Hata detaylarÄ±nÄ± yazdÄ±rma
                 if hasattr(e, 'response') and e.response is not None:
                     print(f"  HTTP Status Code: {e.response.status_code}")
                     try:
                         print(f"  Response Body: {e.response.json()}")
                     except:
                         print(f"  Response Body: {e.response.text}")

            except Exception as e:
                print(f"âŒ Sosyal medya analizi genel hata: {e}")

        else:
            print("âš ï¸ Twitter istemcisi kullanÄ±lamiyor. Sosyal medya analizi atlandÄ±.")


        return signals

    def _check_pump_dump_keywords(self, tweets):
        """Pump&dump anahtar kelimelerini kontrol et"""
        pump_keywords = [
            "moon", "ðŸš€", "pump", "easy money", "quick profit",
            "100x", "1000x", "get rich", "diamond hands", "ape in"
        ]

        # tweets bir liste, her Ã¶ÄŸe bir Tweet nesnesi
        tweet_texts = " ".join([tweet.text.lower() for tweet in tweets])
        return any(keyword in tweet_texts for keyword in pump_keywords)

    def _detect_bot_activity(self, tweets):
        """Bot aktivitesi tespit et"""
        if len(tweets) < 5: # Bot aktivitesi iÃ§in minimum tweet sayÄ±sÄ±
            return False

        # AynÄ± metinlerin tekrarÄ±
        tweet_texts = [tweet.text.strip() for tweet in tweets] # BaÅŸÄ±ndaki ve sonundaki boÅŸluklarÄ± kaldÄ±r
        unique_tweets = set(tweet_texts)

        # %50'den fazlasÄ± aynÄ±ysa bot aktivitesi olabilir (daha dÃ¼ÅŸÃ¼k eÅŸik)
        is_bot = len(unique_tweets) / len(tweet_texts) < 0.5
        if is_bot:
            print("  Bot aktivitesi ÅŸÃ¼phesi: Tekrarlayan tweetler tespit edildi.")
        return is_bot


    def _calculate_sentiment(self, tweets):
        """Basit sentiment analizi"""
        positive_words = ["good", "great", "amazing", "bullish", "buy", "long", "holding", "hodl", "strong"]
        negative_words = ["scam", "rug", "fake", "avoid", "dump", "short", "sell", "warning", "risk"]

        total_score = 0
        if not tweets:
            return 0.5 # Tweet yoksa nÃ¶tr

        for tweet in tweets:
            text = tweet.text.lower()
            pos_count = sum(word in text for word in positive_words)
            neg_count = sum(word in text for word in negative_words)

            # Sentiment skorunu hesapla (basit oran)
            score = 0.5 # Default nÃ¶tr
            if pos_count > neg_count:
                score = 1.0 # Pozitif
            elif neg_count > pos_count:
                score = 0.0 # Negatif
            # EÄŸer pos_count == neg_count ise skor 0.5 kalÄ±r (nÃ¶tr)

            total_score += score

        # Ortalama sentiment
        average_sentiment = total_score / len(tweets)
        return average_sentiment

# utils/code_analysis.py iÃ§eriÄŸi
class CodeAnalyzer:
    def __init__(self):
        self.risk_patterns = {
            "mint_function": [
                r"function\s+mint\s*\(",
                r"function\s+_mint\s*\(",
                r"\.mint\s*\(",
                r"mintTo\s*\("
            ],
            "ownership_issues": [
                r"onlyOwner",
                r"_owner\s*=",
                r"transferOwnership",
                r"renounceOwnership"
            ],
            "liquidity_removal": [
                r"removeLiquidity",
                r"withdraw.*Liquidity",
                r"emergencyWithdraw"
            ],
            "tax_functions": [
                r"buyTax\s*=",
                r"sellTax\s*=",
                r"setTax",
                r"updateTax"
            ],
            "blacklist_functions": [
                r"blacklist",
                r"addBot",
                r"removeBot",
                r"setBot"
            ],
            "pause_functions": [
                r"pause\s*\(",
                r"unpause\s*\(",
                r"setPaused"
            ]
        }

    def analyze_contract_code(self, address, source_code=""):
        """Kontrat kodunu analiz ederek risk bayraklarÄ±nÄ± dÃ¶ndÃ¼r"""
        flags = []

        if not source_code:
            flags.append("âš ï¸ Kontrat kodu doÄŸrulanamadÄ±")
            return flags

        source_code = source_code.lower()

        # Mint fonksiyonu kontrolÃ¼
        if self._check_pattern_exists(source_code, self.risk_patterns["mint_function"]):
            flags.append("ðŸ”´ Mint fonksiyonu tespit edildi")

        # Ownership kontrolleri
        if self._check_pattern_exists(source_code, self.risk_patterns["ownership_issues"]):
            # Not: Ownership renounce kontrolÃ¼ iÃ§in onchain veriye ihtiyaÃ§ var, bu fonksiyon sadece kod patternlerine bakar.
            # OnchainChecker'daki renounce bilgisini burada doÄŸrudan kullanamayÄ±z.
            # Basitlik iÃ§in sadece pattern varlÄ±ÄŸÄ±nÄ± kontrol ediyoruz.
             flags.append("ðŸ”´ Ownership patternleri tespit edildi")


        # Likidite Ã§ekme fonksiyonlarÄ±
        if self._check_pattern_exists(source_code, self.risk_patterns["liquidity_removal"]):
            flags.append("ðŸ”´ Likidite Ã§ekme fonksiyonu var")

        # Vergi fonksiyonlarÄ±
        if self._check_pattern_exists(source_code, self.risk_patterns["tax_functions"]):
            tax_rate = self._extract_tax_rate(source_code)
            if tax_rate > 10:
                flags.append(f"ðŸ”´ YÃ¼ksek vergi oranÄ±: %{tax_rate}")

        # Blacklist fonksiyonlarÄ±
        if self._check_pattern_exists(source_code, self.risk_patterns["blacklist_functions"]):
            flags.append("ðŸ”´ Blacklist fonksiyonu tespit edildi")

        # Pause fonksiyonlarÄ±
        if self._check_pattern_exists(source_code, self.risk_patterns["pause_functions"]):
            flags.append("ðŸ”´ Trading durdurma fonksiyonu var")

        # Honeypot kontrolleri
        if self._check_honeypot_patterns(source_code):
            flags.append("ðŸ”´ Honeypot kalÄ±plarÄ± tespit edildi")

        if not flags:
            flags.append("âœ… Kod analizi temiz")

        return flags

    def _check_pattern_exists(self, code, patterns):
        """Verilen pattern'lerden herhangi biri var mÄ± kontrol et"""
        for pattern in patterns:
            if re.search(pattern, code, re.IGNORECASE):
                return True
        return False

    def _check_ownership_renounced(self, code):
        """Ownership renounce edilmiÅŸ mi kontrol et"""
        # Bu fonksiyon CodeAnalyzer iÃ§inde kullanÄ±lÄ±yordu ancak on-chain veriye ihtiyaÃ§ duyar.
        # Sadece kod patternlerine bakarak renounce kontrolÃ¼ yapmak zordur.
        # Bu nedenle bu fonksiyon burada tam doÄŸru Ã§alÄ±ÅŸmayabilir.
        # OnchainChecker'daki renounced bilgisi daha gÃ¼venilirdir.
        renounce_patterns = [
            r"renounceownership\s*\(\s*\)",
            r"_owner\s*=.*0x0+",
            r"owner.*=.*address\(0\)"
        ]
        return self._check_pattern_exists(code, renounce_patterns) # Sadece pattern var mÄ± diye bakar


    def _extract_tax_rate(self, code):
        """Kod iÃ§inden vergi oranÄ±nÄ± Ã§Ä±karma"""
        tax_patterns = [
            r"buytax\s*=\s*(\d+)",
            r"selltax\s*=\s*(\d+)",
            r"tax.*=\s*(\d+)"
        ]

        max_tax = 0
        for pattern in tax_patterns:
            matches = re.findall(pattern, code, re.IGNORECASE)
            for match in matches:
                try:
                    max_tax = max(max_tax, int(match))
                except ValueError:
                    print(f"  _extract_tax_rate ValueError: {match} int'e Ã§evrilemedi.")
                    continue # HatalÄ± veriyi atla

        return max_tax

    def _check_honeypot_patterns(self, code):
        """Honeypot kalÄ±plarÄ±nÄ± kontrol et"""
        honeypot_patterns = [
            r"require.*balances\[.*\]\s*>=",  # Balance kontrolÃ¼
            r"if.*balances\[.*\].*return false",  # Transfer engelleme
            r"mapping.*bool.*isexcluded",  # Exclusion mapping
            r"tradingopen\s*=\s*false"  # Trading kapalÄ±
        ]
        return self._check_pattern_exists(code, honeypot_patterns)

# utils/feature_engineering.py iÃ§eriÄŸi
class FeatureEngineer:
    def build_features(self, onchain_data, social_data=None):
        """Ham verilerden ML modeli iÃ§in Ã¶zellikler oluÅŸtur"""

        # Temel Ã¶zellikler
        features = [
            onchain_data.get("holder_concentration", 75),
            int(onchain_data.get("mint_function", True)),
            int(onchain_data.get("lp_locked", False)),
            onchain_data.get("max_tax", 25),
            int(onchain_data.get("owner_renounced", False)),
            int(onchain_data.get("contract_verified", False))
        ]

        # Kontrat yaÅŸÄ± (gÃ¼n cinsinden)
        creation_date = onchain_data.get("creation_date", 0)
        if creation_date > 0:
            age_days = (datetime.now().timestamp() - creation_date) / 86400
            features.append(min(age_days, 365))  # Max 1 yÄ±l
        else:
            features.append(0)

        # Holder sayÄ±sÄ± (log Ã¶lÃ§eÄŸi)
        holder_count = onchain_data.get("holder_count", 1)
        features.append(np.log10(max(holder_count, 1)))

        # Sosyal medya Ã¶zellikleri
        if social_data:
            features.extend([
                social_data.get("twitter_mentions", 0),
                social_data.get("sentiment_score", 0.5) * 100,  # 0-100 Ã¶lÃ§eÄŸi
                int(social_data.get("bot_activity", False)),
                int(social_data.get("pump_dump_keywords", False))
            ])
        else:
            features.extend([0, 50, 0, 0])  # Default deÄŸerler

        return np.array(features)

    def get_feature_names(self):
        """Ã–zellik isimlerini dÃ¶ndÃ¼r"""
        return [
            "holder_concentration",
            "mint_function",
            "lp_locked",
            "max_tax",
            "owner_renounced",
            "contract_verified",
            "age_days",
            "log_holder_count",
            "twitter_mentions",
            "sentiment_score",
            "bot_activity",
            "pump_dump_keywords"
        ]

# Ana ScamDetector sÄ±nÄ±fÄ± (main.py'nin Ã¶nceki iÃ§eriÄŸi)
class ScamDetector:
    def __init__(self):
        self.onchain_checker = OnchainChecker()
        self.social_analyzer = SocialAnalyzer()
        self.code_analyzer = CodeAnalyzer()
        self.feature_engineer = FeatureEngineer()

        # Modeli yÃ¼kle (placeholder)
        # GerÃ§ek bir model eÄŸitildiÄŸinde bu kÄ±sÄ±m kullanÄ±lacak
        self.model = None
        # model_path = "models/scam_model.pkl"
        # if os.path.exists(model_path):
        #     try:
        #         self.model = joblib.load(model_path)
        #         print("âœ… Model baÅŸarÄ±yla yÃ¼klendi.")
        #     except Exception as e:
        #         print(f"âŒ Model yÃ¼kleme hatasÄ±: {e}")
        #         self.model = None
        # else:
        #     print("âš ï¸ Model bulunamadÄ±. Risk skoru placeholder olacaktÄ±r.")


    def analyze_contract(self, contract_address, token_name=""):
        """Ana analiz fonksiyonu"""
        print(f"ðŸ” Analiz baÅŸlÄ±yor: {contract_address}")

        try:
            # 1. Onchain verilerini al
            print("ðŸ“Š Onchain veriler alÄ±nÄ±yor...")
            onchain_features = self.onchain_checker.get_contract_features(contract_address)

            # 2. Sosyal medya analizi (opsiyonel)
            social_features = None
            # Sosyal medya analizi iÃ§in token_name veya contract_address gerekiyor
            if token_name or contract_address:
                 print("ðŸ¦ Sosyal medya analizi yapÄ±lÄ±yor...")
                 # SocialAnalyzer analyze_social_signals method expects token_name and contract_address
                 social_features = self.social_analyzer.analyze_social_signals(token_name, contract_address)


            # 3. Kod analizi
            print("ðŸ’» Kontrat kodu analiz ediliyor...")
            # get_contract_source is a private method, access it carefully or make it public if needed
            # Dummy checker durumunda source_code boÅŸ olacaktÄ±r, bu CodeAnalyzer tarafÄ±ndan iÅŸlenecektir.
            source_code = self.onchain_checker._get_contract_source(contract_address).get("source_code", "")
            code_flags = self.code_analyzer.analyze_contract_code(contract_address, source_code)

            # 4. ML Ã¶zelliklerini oluÅŸtur
            print("âš™ï¸ Ã–zellik mÃ¼hendisliÄŸi yapÄ±lÄ±yor...")
            features = self.feature_engineer.build_features(onchain_features, social_features)

            # 5. Risk skorunu hesapla (placeholder veya model ile)
            risk_score = self._calculate_placeholder_risk(onchain_features, code_flags) # Default placeholder
            ml_prediction = "Tahmin Yok (Model Yok)"


            if self.model is not None:
                try:
                    # Model tahminini yap
                    # Modelin beklediÄŸi input formatÄ±na dikkat edin (genellikle 2D array)
                    features_2d = np.array([features]) # Model predict tek Ã¶rnek iÃ§in 2D array bekler
                    risk_probability = self.model.predict_proba(features_2d)[0][1]
                    risk_score = round(risk_probability * 100, 2)
                    ml_prediction = "Scam" if risk_probability > 0.7 else "GÃ¼venli" if risk_probability < 0.3 else "Dikkatli ol"
                    print(f"âœ… Model tahmini yapÄ±ldÄ±: Risk OlasÄ±lÄ±ÄŸÄ± = {risk_probability:.2f}, Tahmin: {ml_prediction}")
                except Exception as e:
                    print(f"âŒ Model tahmin hatasÄ±: {e}")
                    ml_prediction = f"Model HatasÄ±: {e}"


            # 6. SonuÃ§larÄ± birleÅŸtir
            result = {
                "contract": contract_address,
                "token_name": token_name,
                "risk_score": risk_score,
                "ml_prediction": ml_prediction,
                "red_flags": code_flags,
                "onchain_data": onchain_features,
                "social_data": social_features,
                "analysis_time": datetime.now().strftime("%Y-%m-%d %H:%M:%S"),
                "recommendations": self._generate_recommendations(risk_score, onchain_features, code_flags)
            }

            return result

        except Exception as e:
            print(f"âŒ Analiz hatasÄ±: {e}")
            import traceback
            traceback.print_exc() # Hata detaylarÄ±nÄ± yazdÄ±r

            return {
                "contract": contract_address,
                "risk_score": 100,  # Hata durumunda yÃ¼ksek risk
                "ml_prediction": "Analiz edilemedi",
                "red_flags": [f"âŒ Analiz hatasÄ±: {str(e)}"],
                "error": str(e)
            }

    def _calculate_placeholder_risk(self, onchain_data, code_flags):
        """Basit placeholder risk skoru hesaplama"""
        score = 0

        # On-chain verilere gÃ¶re puanlama
        # onchain_data'nÄ±n boÅŸ olup olmadÄ±ÄŸÄ±nÄ± kontrol edin
        if onchain_data:
            if onchain_data.get("holder_concentration", 75) > 60: score += 20
            if onchain_data.get("mint_function", True): score += 20
            if not onchain_data.get("lp_locked", False): score += 20
            if onchain_data.get("max_tax", 25) > 15: score += 15
            if not onchain_data.get("owner_renounced", False): score += 15
            if not onchain_data.get("contract_verified", False): score += 10
        else:
             score += 50 # On-chain veri alÄ±namÄ±yorsa riskli say


        # Kod analizindeki bayraklara gÃ¶re puanlama
        if "ðŸ”´ Mint fonksiyonu tespit edildi" in code_flags: score += 10
        if "ðŸ”´ Owner haklarÄ± korunuyor" in code_flags: score += 10
        if "ðŸ”´ Likidite Ã§ekme fonksiyonu var" in code_flags: score += 10
        # YÃ¼ksek vergi oranÄ± bayraÄŸÄ± zaten oranÄ± iÃ§eriyor, sadece bayraÄŸÄ±n varlÄ±ÄŸÄ±nÄ± kontrol edelim.
        if any("ðŸ”´ YÃ¼ksek vergi oranÄ±" in flag for flag in code_flags): score += 10
        if "ðŸ”´ Blacklist fonksiyonu tespit edildi" in code_flags: score += 15
        if "ðŸ”´ Trading durdurma fonksiyonu var" in code_flags: score += 15
        if "ðŸ”´ Honeypot kalÄ±plarÄ± tespit edildi" in code_flags: score += 25
        if "âš ï¸ Kontrat kodu doÄŸrulanamadÄ±" in code_flags: score += 20 # DoÄŸrulanamayan kod risklidir.


        return min(score, 100) # Max 100


    def _generate_recommendations(self, risk_score, onchain_data, code_flags):
        """Risk skoruna gÃ¶re Ã¶neriler oluÅŸtur"""
        recommendations = []

        if risk_score >= 80:
            recommendations.append("ðŸš¨ YÃœKÂ­SEK RÄ°SK: Bu token'a yatÄ±rÄ±m yapmayÄ±n!")
            recommendations.append("ðŸ’¡ BaÅŸka projeleri araÅŸtÄ±rÄ±n")
        elif risk_score >= 60:
            recommendations.append("âš ï¸ ORTA RÄ°SK: Ã‡ok dikkatli olun")
            recommendations.append("ðŸ’¡ KÃ¼Ã§Ã¼k miktarla test edin")
            recommendations.append("ðŸ’¡ Exit stratejinizi belirleyin")
        elif risk_score >= 40:
            recommendations.append("ðŸŸ¡ DÃœÂ­ÅžÃœK-ORTA RÄ°SK: AraÅŸtÄ±rma yapÄ±n")
            recommendations.append("ðŸ’¡ Topluluk ve geliÅŸtirici aktivitesini kontrol edin")
        else:
            recommendations.append("âœ… DÃœÂ­ÅžÃœK RÄ°SK: GÃ¶rece gÃ¼venli gÃ¶rÃ¼nÃ¼yor")
            recommendations.append("ðŸ’¡ Yine de kendi araÅŸtÄ±rmanÄ±zÄ± yapÄ±n")

        # Ã–zel durumlar iÃ§in Ã¶neriler (onchain_data'nÄ±n varlÄ±ÄŸÄ±nÄ± kontrol edin)
        if onchain_data:
            if not onchain_data.get("contract_verified", False):
                recommendations.append("ðŸ’¡ Kontrat kodu doÄŸrulanmamÄ±ÅŸ - ekstra dikkat")

            if onchain_data.get("holder_concentration", 0) > 50:
                recommendations.append("ðŸ’¡ Token daÄŸÄ±lÄ±mÄ± centralized - whale riski")

            if not onchain_data.get("owner_renounced", False):
                recommendations.append("ðŸ’¡ Owner haklarÄ± hala aktif - kontrol riski")

        # Code flags'e Ã¶zel Ã¶neriler
        if "ðŸ”´ Mint fonksiyonu tespit edildi" in code_flags:
             recommendations.append("ðŸ’¡ Mint fonksiyonu, token arzÄ±nÄ±n artÄ±rÄ±labileceÄŸi anlamÄ±na gelir - dikkatli olun.")
        if "ðŸ”´ Likidite Ã§ekme fonksiyonu var" in code_flags:
             recommendations.append("ðŸ’¡ Likidite Ã§ekme fonksiyonu rug pull riskini artÄ±rÄ±r.")
        if any("ðŸ”´ YÃ¼ksek vergi oranÄ±" in flag for flag in code_flags):
             recommendations.append("ðŸ’¡ YÃ¼ksek vergi oranlarÄ± alÄ±m satÄ±m maliyetini artÄ±rÄ±r ve scam gÃ¶stergesi olabilir.")
        if "ðŸ”´ Blacklist fonksiyonu tespit edildi" in code_flags:
             recommendations.append("ðŸ’¡ Blacklist fonksiyonu, cÃ¼zdanlarÄ±n alÄ±m satÄ±m yapmasÄ±nÄ± engelleyebilir.")
        if "ðŸ”´ Trading durdurma fonksiyonu var" in code_flags:
             recommendations.append("ðŸ’¡ Trading durdurma fonksiyonu, token'Ä±n satÄ±lamaz hale gelmesine neden olabilir.")
        if "ðŸ”´ Honeypot kalÄ±plarÄ± tespit edildi" in code_flags:
             recommendations.append("ðŸ’¡ Honeypot kalÄ±plarÄ±, token'Ä± alÄ±p satmanÄ±zÄ± engelleyebilir (sadece alÄ±m yapÄ±labilir).")


        return recommendations


# Gradio ArayÃ¼zÃ¼
def analyze_token_interface(contract_address, token_name):
    """Gradio arayÃ¼zÃ¼ iÃ§in Ã§aÄŸrÄ±lacak fonksiyon"""
    detector = ScamDetector()
    result = detector.analyze_contract(contract_address, token_name)

    # Gradio Ã§Ä±ktÄ±sÄ± iÃ§in formatlama
    output_text = f"## ðŸ¤– Kripto Token Risk Analiz Botu SonuÃ§larÄ±\n\n"
    output_text += f"**ðŸ“§ Kontrat:** {result.get('contract', 'N/A')}\n"
    output_text += f"**ðŸ“› Token AdÄ±:** {result.get('token_name', 'N/A')}\n"

    # Risk skoruna gÃ¶re renkli Ã§Ä±ktÄ±
    risk_score = result.get('risk_score', 'N/A')
    if isinstance(risk_score, (int, float)):
        if risk_score >= 80:
            output_text += f"**ðŸŽ¯ Risk Skoru:** <span style='color: red; font-weight: bold;'>{risk_score}% (YÃ¼ksek Risk)</span>\n"
        elif risk_score >= 60:
            output_text += f"**ðŸŽ¯ Risk Skoru:** <span style='color: orange; font-weight: bold;'>{risk_score}% (Orta Risk)</span>\n"
        elif risk_score >= 40:
             output_text += f"**ðŸŽ¯ Risk Skoru:** <span style='color: yellow; font-weight: bold;'>{risk_score}% (DÃ¼ÅŸÃ¼k-Orta Risk)</span>\n"
        else:
            output_text += f"**ðŸŽ¯ Risk Skoru:** <span style='color: green; font-weight: bold;'>{risk_score}% (DÃ¼ÅŸÃ¼k Risk)</span>\n"
    else:
         output_text += f"**ðŸŽ¯ Risk Skoru:** {risk_score}\n"


    output_text += f"**ðŸ¤– ML Tahmini:** {result.get('ml_prediction', 'N/A')}\n"
    output_text += f"**ðŸ“… Analiz ZamanÄ±:** {result.get('analysis_time', 'Bilinmiyor')}\n\n"

    output_text += f"### ðŸš© Risk BayraklarÄ±:\n"
    if result.get('red_flags'):
        # Bayraklara gÃ¶re renkli Ã§Ä±ktÄ±
        for flag in result['red_flags']:
            if "ðŸ”´" in flag:
                output_text += f"- <span style='color: red;'>{flag}</span>\n"
            elif "âš ï¸" in flag:
                output_text += f"- <span style='color: orange;'>{flag}</span>\n"
            elif "âœ…" in flag:
                 output_text += f"- <span style='color: green;'>{flag}</span>\n"
            else:
                output_text += f"- {flag}\n"

    else:
        output_text += "- Risk bayraÄŸÄ± bulunamadÄ±.\n"
    output_text += "\n"

    output_text += f"### ðŸ’¡ Ã–neriler:\n"
    if result.get('recommendations'):
        for rec in result['recommendations']:
            # Ã–nerilere gÃ¶re renkli Ã§Ä±ktÄ±
            if "ðŸš¨ YÃœKÂ­SEK RÄ°SK" in rec:
                output_text += f"- <span style='color: red; font-weight: bold;'>{rec}</span>\n"
            elif "âš ï¸ ORTA RÄ°SK" in rec:
                output_text += f"- <span style='color: orange; font-weight: bold;'>{rec}</span>\n"
            elif "ðŸŸ¡ DÃœÂ­ÅžÃœK-ORTA RÄ°SK" in rec:
                 output_text += f"- <span style='color: yellow; font-weight: bold;'>{rec}</span>\n"
            elif "âœ… DÃœÂ­ÅžÃœK RÄ°SK" in rec:
                 output_text += f"- <span style='color: green; font-weight: bold;'>{rec}</span>\n"
            else:
                output_text += f"- {rec}\n"
    else:
        output_text += "- Ã–neri bulunamadÄ±.\n"
    output_text += "\n"

    if result.get('onchain_data'):
        output_text += f"### ðŸ“Š Onchain Veriler:\n"
        data = result['onchain_data']
        output_text += f"- Holder Konsantrasyonu: %{data.get('holder_concentration', 0):.1f}\n"
        output_text += f"- Mint Fonksiyonu: {'Var' if data.get('mint_function') else 'Yok'}\n"
        output_text += f"- LP Kilidi: {'Var' if data.get('lp_locked') else 'Yok'}\n"
        output_text += f"- Maksimum Tax: %{data.get('max_tax', 0)}\n"
        output_text += f"- Owner Renounced: {'Evet' if data.get('owner_renounced') else 'HayÄ±r'}\n"
        output_text += f"- Kontrat DoÄŸrulanmÄ±ÅŸ: {'Evet' if data.get('contract_verified') else 'HayÄ±r'}\n"
        # Timestamp'i okunabilir tarihe Ã§evirelim
        creation_date_ts = data.get('creation_date', 0)
        if creation_date_ts > 0:
             creation_date_str = datetime.fromtimestamp(creation_date_ts).strftime("%Y-%m-%d %H:%M:%S")
             output_text += f"- Kontrat OluÅŸturulma Tarihi: {creation_date_str}\n"
        else:
             output_text += f"- Kontrat OluÅŸturulma Tarihi: Bilinmiyor\n"

        output_text += f"- Toplam Arz: {data.get('total_supply', 0)}\n"
        output_text += f"- Holder SayÄ±sÄ±: {data.get('holder_count', 0)}\n"
        output_text += "\n"


    if result.get('social_data'):
        output_text += f"### ðŸ¦ Sosyal Medya Verileri:\n"
        social_data = result['social_data']
        output_text += f"- Twitter Mention SayÄ±sÄ± (Son 24s): {social_data.get('twitter_mentions', 0)}\n"
        output_text += f"- Sentiment Skoru (0-100): {social_data.get('sentiment_score', 0):.1f}\n"
        output_text += f"- Bot Aktivitesi Tespit Edildi: {'Evet' if social_data.get('bot_activity') else 'HayÄ±r'}\n"
        output_text += f"- Pump/Dump Anahtar Kelimeleri: {'Var' if social_data.get('pump_dump_keywords') else 'Yok'}\n"
        output_text += "\n"

    if result.get('error'):
         output_text += f"### âŒ Hata:\n"
         output_text += f"<span style='color: red;'>{result['error']}</span>\n"


    return output_text

# Gradio ArayÃ¼zÃ¼nÃ¼ OluÅŸtur
if __name__ == "__main__":
    interface = gr.Interface(
        fn=analyze_token_interface,
        inputs=[
            gr.Textbox(label="Kontrat Adresi (0x...)"),
            gr.Textbox(label="Token AdÄ± (Opsiyonel, Sosyal Medya Analizi Ä°Ã§in)")
        ],
        outputs=gr.Markdown(label="Analiz SonuÃ§larÄ±"),
        title="ðŸ¤– Kripto Token Risk Analiz Botu",
        description="Bir kontrat adresini girerek token'Ä±n on-chain ve sosyal medya risklerini analiz edin."
    )

    # ArayÃ¼zÃ¼ baÅŸlat
    # Colab'da Ã§alÄ±ÅŸtÄ±rmak iÃ§in share=True kullanÄ±n
    interface.launch(share=True)
```

```python
%%writefile utils/onchain_checker.py
import requests
from web3 import Web3
from config import Config
import json
import time

class OnchainChecker:
    def __init__(self):
        self.w3 = Web3(Web3.HTTPProvider(Config.INFURA_URL))
        self.etherscan_api = Config.ETHERSCAN_API_KEY
        # Check Infura connection
        if not self.w3.is_connected():
            print("âŒ Infura baÄŸlantÄ±sÄ± kurulamadÄ±! INFURA_PROJECT_ID'nizi kontrol edin.")
        else:
            print("âœ… Infura baÄŸlantÄ±sÄ± baÅŸarÄ±lÄ±.")


    def get_contract_features(self, address):
        try:
            # Kontrat adresini doÄŸrula
            if not self.w3.is_address(address):
                raise ValueError("GeÃ§ersiz kontrat adresi")

            address = self.w3.to_checksum_address(address)
            print(f"ðŸ” Analiz edilen adres (checksum): {address}")


            # Token bilgilerini al
            print("-> Token bilgileri alÄ±nÄ±yor...")
            token_info = self._get_token_info(address)
            print(f"<- Token bilgileri: {token_info}")

            print("-> Holder daÄŸÄ±lÄ±mÄ± alÄ±nÄ±yor...")
            holder_data = self._get_holder_distribution(address)
            print(f"<- Holder sayÄ±sÄ±: {len(holder_data)}")


            print("-> Kontrat kaynak kodu alÄ±nÄ±yor...")
            contract_code = self._get_contract_source(address)
            print(f"<- Kontrat doÄŸrulanmÄ±ÅŸ: {contract_code.get('verified', False)}")


            print("-> Likidite kilidi kontrol ediliyor...")
            liquidity_info = self._check_liquidity_lock(address)
            print(f"<- LP kilitli: {liquidity_info.get('locked', False)}")


            print("-> Owner renounce kontrol ediliyor...")
            owner_renounced = self._check_ownership_renounced(address)
            print(f"<- Owner renounce edilmiÅŸ: {owner_renounced}")


            print("-> Kontrat oluÅŸturulma tarihi alÄ±nÄ±yor...")
            creation_date = self._get_creation_date(address)
            print(f"<- OluÅŸturulma tarihi (timestamp): {creation_date}")


            return {
                "holder_concentration": self._calculate_concentration(holder_data),
                "mint_function": self._check_mint_function(contract_code),
                "lp_locked": liquidity_info.get("locked", False),
                "max_tax": self._extract_max_tax(contract_code),
                "owner_renounced": owner_renounced,
                "contract_verified": contract_code.get("verified", False),
                "creation_date": creation_date,
                "total_supply": token_info.get("total_supply", 0),
                "holder_count": len(holder_data)
            }
        except Exception as e:
            print(f"âŒ OnchainChecker hatasÄ±: {e}")
            return self._get_default_features()

    def _get_token_info(self, address):
        """Token temel bilgilerini al"""
        try:
            # ERC-20 ABI (sadece gerekli fonksiyonlar)
            erc20_abi = [
                {
                    "constant": True,
                    "inputs": [],
                    "name": "totalSupply",
                    "outputs": [{"name": "", "type": "uint256"}],
                    "type": "function"
                }
            ]

            contract = self.w3.eth.contract(address=address, abi=erc20_abi)
            total_supply = contract.functions.totalSupply().call()

            return {"total_supply": total_supply}
        except Exception as e:
            print(f"  _get_token_info hatasÄ±: {e}")
            return {"total_supply": 0}

    def _get_holder_distribution(self, address):
        """Holder daÄŸÄ±lÄ±mÄ±nÄ± Etherscan'den al"""
        try:
            url = f"https://api.etherscan.io/api"
            params = {
                "module": "token",
                "action": "tokenholderlist",
                "contractaddress": address,
                "page": 1,
                "offset": 100,
                "apikey": self.etherscan_api
            }
            # requests handles urlencoding parameters automatically
            print(f"  Etherscan API Ã§aÄŸrÄ±sÄ± (tokenholderlist): {url}?module=token&action=tokenholderlist&contractaddress={address}&page=1&offset=100&apikey=...")

            response = requests.get(url, params=params, timeout=15) # Timeout artÄ±rÄ±ldÄ±
            response.raise_for_status() # HTTP hatalarÄ±nÄ± yakala
            data = response.json()
            print(f"  Etherscan API yanÄ±tÄ± (tokenholderlist status): {data.get('status')}, message: {data.get('message')}")


            if data.get("status") == "1":
                return data.get("result", [])
            else:
                print(f"  Etherscan API'den holder bilgisi alÄ±namadÄ±: {data.get('message', 'Bilinmeyen hata')}")
                return []
        except requests.exceptions.RequestException as e:
            print(f"  _get_holder_distribution istek hatasÄ±: {e}")
            return []
        except Exception as e:
            print(f"  _get_holder_distribution genel hata: {e}")
            return []


    def _calculate_concentration(self, holders):
        """Top 10 holder konsantrasyonunu hesapla"""
        if not holders:
            return 100  # Veri yoksa risk yÃ¼ksek

        try:
            # Ä°lk 10 holder'Ä±n toplam token yÃ¼zdesi
            # TokenHolderQuantity string gelebilir, float'a Ã§evirirken hata olabilir
            top_10_balance = 0
            for h in holders[:10]:
                try:
                    top_10_balance += float(h.get("TokenHolderQuantity", 0))
                except ValueError:
                    print(f"  _calculate_concentration ValueError: {h.get('TokenHolderQuantity')} float'a Ã§evrilemedi.")
                    continue # HatalÄ± veriyi atla

            total_supply = 0
            for h in holders:
                 try:
                    total_supply += float(h.get("TokenHolderQuantity", 0))
                 except ValueError:
                     print(f"  _calculate_concentration ValueError: {h.get('TokenHolderQuantity')} float'a Ã§evrilemedi.")
                     continue # HatalÄ± veriyi atla


            if total_supply == 0:
                return 100

            concentration = (top_10_balance / total_supply) * 100
            return min(concentration, 100)  # Max %100
        except Exception as e:
            print(f"  _calculate_concentration genel hata: {e}")
            return 75  # Default deÄŸer

    def _get_contract_source(self, address):
        """Kontrat kaynak kodunu al"""
        try:
            url = f"https://api.etherscan.io/api"
            params = {
                "module": "contract",
                "action": "getsourcecode",
                "address": address,
                "apikey": self.etherscan_api
            }
            # requests handles urlencoding parameters automatically
            print(f"  Etherscan API Ã§aÄŸrÄ±sÄ± (getsourcecode): {url}?module=contract&action=getsourcecode&address={address}&apikey=...")


            response = requests.get(url, params=params, timeout=15) # Timeout artÄ±rÄ±ldÄ±
            response.raise_for_status() # HTTP hatalarÄ±nÄ± yakala
            data = response.json()
            print(f"  Etherscan API yanÄ±tÄ± (getsourcecode status): {data.get('status')}, message: {data.get('message')}")


            if data.get("status") == "1" and data.get("result"):
                source_code = data["result"][0].get("SourceCode", "")
                return {
                    "source_code": source_code,
                    "verified": len(source_code) > 0 and source_code != "Contract source code not verified" # DoÄŸrulama mesajÄ±nÄ± da kontrol et
                }
            else:
                 print(f"  Etherscan API'den kaynak kodu alÄ±namadÄ±: {data.get('message', 'Bilinmeyen hata')}")
                 return {"source_code": "", "verified": False}

        except requests.exceptions.RequestException as e:
            print(f"  _get_contract_source istek hatasÄ±: {e}")
            return {"source_code": "", "verified": False}
        except Exception as e:
            print(f"  _get_contract_source genel hata: {e}")
            return {"source_code": "", "verified": False}


    def _check_mint_function(self, contract_data):
        """Mint fonksiyonu var mÄ± kontrol et"""
        source_code = contract_data.get("source_code", "").lower()
        mint_keywords = ["function mint", "mint(", "_mint(", "mintto", "mint_"]
        return any(keyword in source_code for keyword in mint_keywords)

    def _extract_max_tax(self, contract_data):
        """Maksimum tax oranÄ±nÄ± bulmaya Ã§alÄ±ÅŸ"""
        source_code = contract_data.get("source_code", "").lower()

        # YaygÄ±n tax variable isimleri
        tax_patterns = ["buytax", "selltax", "tax", "fee"]

        import re
        max_tax = 0
        for pattern in tax_patterns:
            # uint256 public buyTax = 5; gibi patternleri ara
            regex = rf"{pattern}\s*=\s*(\d+)"
            matches = re.findall(regex, source_code)
            for match in matches:
                 try:
                     max_tax = max(max_tax, int(match))
                 except ValueError:
                     print(f"  _extract_max_tax ValueError: {match} int'e Ã§evrilemedi.")
                     continue # HatalÄ± veriyi atla

        return max_tax # Default deÄŸer


    def _check_ownership_renounced(self, address):
        """Owner renounce edilmiÅŸ mi kontrol et"""
        try:
            # Owner fonksiyonu Ã§aÄŸÄ±r
            owner_abi = [{
                "constant": True,
                "inputs": [],
                "name": "owner",
                "outputs": [{"name": "", "type": "address"}],
                "type": "function"
            }]

            contract = self.w3.eth.contract(address=address, abi=owner_abi)
            owner = contract.functions.owner().call()

            # 0x000...000 adresine sahipse renounce edilmiÅŸ
            return owner == "0x0000000000000000000000000000000000000000"
        except Exception as e:
            print(f"  _check_ownership_renounced hatasÄ±: {e}")
            return False

    def _check_liquidity_lock(self, address):
        """Likidite kilidi kontrol et (basit versiyon)"""
        # Bu gerÃ§ek implementasyonda Uniswap/PancakeSwap pool kontratlarÄ±nÄ± kontrol etmeli
        # Etherscan API'sinde doÄŸrudan likidite kilidi bilgisi yok, bu daha karmaÅŸÄ±k bir analiz gerektirir.
        # Åžimdilik sabit deÄŸer dÃ¶ndÃ¼rÃ¼yoruz.
        print("  _check_liquidity_lock: Basit kontrol, her zaman False dÃ¶ndÃ¼rÃ¼yor.")
        return {"locked": False, "lock_time": 0}

    def _get_creation_date(self, address):
        """Kontrat oluÅŸturulma tarihini al"""
        try:
            url = f"https://api.etherscan.io/api"
            params = {
                "module": "account",
                "action": "txlist",
                "address": address,
                "startblock": 0,
                "endblock": 99999999,
                "page": 1,
                "offset": 1,
                "sort": "asc",
                "apikey": self.etherscan_api
            }
            # requests handles urlencoding parameters automatically
            print(f"  Etherscan API Ã§aÄŸrÄ±sÄ± (txlist): {url}?module=account&action=txlist&address={address}&startblock=0&endblock=99999999&page=1&offset=1&sort=asc&apikey=...")


            response = requests.get(url, params=params, timeout=15) # Timeout artÄ±rÄ±ldÄ±
            response.raise_for_status() # HTTP hatalarÄ±nÄ± yakala
            data = response.json()
            print(f"  Etherscan API yanÄ±tÄ± (txlist status): {data.get('status')}, message: {data.get('message')}")


            if data.get("status") == "1" and data.get("result"):
                timestamp = int(data["result"][0].get("timeStamp", 0))
                return timestamp
            else:
                print(f"  Etherscan API'den oluÅŸturulma tarihi alÄ±namadÄ±: {data.get('message', 'Bilinmeyen hata')}")
                return 0
        except requests.exceptions.RequestException as e:
            print(f"  _get_creation_date istek hatasÄ±: {e}")
            return 0
        except Exception as e:
            print(f"  _get_creation_date genel hata: {e}")
            return 0


    def _get_default_features(self):
        """Hata durumunda default deÄŸerler"""
        return {
            "holder_concentration": 75,
            "mint_function": True,
            "lp_locked": False,
            "max_tax": 25,
            "owner_renounced": False,
            "contract_verified": False,
            "creation_date": 0,
            "total_supply": 0,
            "holder_count": 0
        }
```

```python
%%writefile utils/social_analysis.py
import requests
import tweepy
from config import Config
from datetime import datetime, timedelta

class SocialAnalyzer:
    def __init__(self):
        if Config.TWITTER_BEARER_TOKEN:
            try:
                self.twitter_client = tweepy.Client(bearer_token=Config.TWITTER_BEARER_TOKEN)
                # Verify credentials - basic check
                # self.twitter_client.get_me() # Bu kotaya dahil olabilir, dikkatli kullanÄ±n
                print("âœ… Twitter istemcisi baÅŸlatÄ±ldÄ±.")
            except Exception as e:
                 print(f"âŒ Twitter istemcisi baÅŸlatÄ±lÄ±rken hata oluÅŸtu: {e}")
                 self.twitter_client = None
        else:
            print("âš ï¸ TWITTER_BEARER_TOKEN ayarlanmamÄ±ÅŸ. Sosyal medya analizi yapÄ±lamayacak.")
            self.twitter_client = None


    def analyze_social_signals(self, token_name, contract_address):
        """Token iÃ§in sosyal medya sinyallerini analiz et"""
        signals = {
            "twitter_mentions": 0,
            "sentiment_score": 0.5,  # 0-1 arasÄ±
            "bot_activity": False,
            "pump_dump_keywords": False
        }

        if self.twitter_client:
            try:
                # Sorgu oluÅŸturma - Hata ayÄ±klama iÃ§in basÄ±ldÄ±
                # Tweet sorgusu iÃ§in kurallar: https://developer.twitter.com/en/docs/twitter-api/tweets/search/integrate/build-a-query
                # Contract adresi yerine sadece token adÄ± veya kÄ±saltmasÄ± daha iyi olabilir
                # Ã‡ok kÄ±sa contract adresi parÃ§asÄ± ([:10]) geÃ§ersiz sorguya neden olabilir
                # Token adÄ± boÅŸsa veya Ã§ok kÄ±saysa da sorun olabilir.
                query_parts = []
                # Check if token_name is provided and is reasonably long
                if token_name and len(token_name.strip()) > 2: # Use token name if provided and > 2 chars
                    query_parts.append(f'"{token_name.strip()}"')
                # Add contract address only if token_name is not used or is short, and address is valid length
                # Using the full address in query might be too specific, maybe just the beginning?
                # Or perhaps only use contract address if no token name is given?
                # Let's prioritize token name, then fall back to part of address if name is not useful.
                elif contract_address and len(contract_address) > 10: # Check contract address length
                     query_parts.append(f'{contract_address[:10]}') # Use first 10 chars


                if not query_parts:
                     print("âš ï¸ Twitter sorgusu iÃ§in yeterli bilgi (token adÄ± veya kontrat adresi) yok.")
                     return signals # BoÅŸ sinyaller dÃ¶ndÃ¼r

                # Combine query parts with OR, ensure no leading/trailing OR if only one part
                query = " OR ".join(query_parts)
                final_query = f"{query} -is:retweet lang:en" # Retweetleri hariÃ§ tut

                print(f"  Twitter API sorgusu: {final_query}")

                # Son 24 saat iÃ§inde token hakkÄ±nda tweetleri ara
                # start_time = datetime.utcnow() - timedelta(hours=24)
                # start_time_str = start_time.isoformat("T") + "Z" # RFC3339 formatÄ±


                tweets = self.twitter_client.search_recent_tweets(
                    query=final_query,
                    max_results=10, # Test iÃ§in daha az sonuÃ§ Ã§ek
                    # start_time=start_time_str, # Son 24 saat
                    tweet_fields=["created_at", "author_id", "public_metrics", "text"] # text'i de al
                )

                if tweets and tweets.data:
                    print(f"  Twitter API {len(tweets.data)} tweet buldu.")
                    signals["twitter_mentions"] = len(tweets.data)
                    signals["pump_dump_keywords"] = self._check_pump_dump_keywords(tweets.data)
                    signals["bot_activity"] = self._detect_bot_activity(tweets.data)
                    signals["sentiment_score"] = self._calculate_sentiment(tweets.data)
                else:
                    print("  Twitter API tweet bulamadÄ±.")
                    signals["twitter_mentions"] = 0


            except tweepy.errors.TweepyException as e:
                 print(f"âŒ Twitter API hatasÄ± (TweepyException): {e}")
                 # Hata detaylarÄ±nÄ± yazdÄ±rma
                 if hasattr(e, 'response') and e.response is not None:
                     print(f"  HTTP Status Code: {e.response.status_code}")
                     try:
                         print(f"  Response Body: {e.response.json()}")
                     except:
                         print(f"  Response Body: {e.response.text}")

            except Exception as e:
                print(f"âŒ Sosyal medya analizi genel hata: {e}")

        else:
            print("âš ï¸ Twitter istemcisi kullanÄ±lamiyor. Sosyal medya analizi atlandÄ±.")


        return signals

    def _check_pump_dump_keywords(self, tweets):
        """Pump&dump anahtar kelimelerini kontrol et"""
        pump_keywords = [
            "moon", "ðŸš€", "pump", "easy money", "quick profit",
            "100x", "1000x", "get rich", "diamond hands", "ape in"
        ]

        # tweets bir liste, her Ã¶ÄŸe bir Tweet nesnesi
        tweet_texts = " ".join([tweet.text.lower() for tweet in tweets])
        return any(keyword in tweet_texts for keyword in pump_keywords)

    def _detect_bot_activity(self, tweets):
        """Bot aktivitesi tespit et"""
        if len(tweets) < 5: # Bot aktivitesi iÃ§in minimum tweet sayÄ±sÄ±
            return False

        # AynÄ± metinlerin tekrarÄ±
        tweet_texts = [tweet.text.strip() for tweet in tweets] # BaÅŸÄ±ndaki ve sonundaki boÅŸluklarÄ± kaldÄ±r
        unique_tweets = set(tweet_texts)

        # %50'den fazlasÄ± aynÄ±ysa bot aktivitesi olabilir (daha dÃ¼ÅŸÃ¼k eÅŸik)
        is_bot = len(unique_tweets) / len(tweet_texts) < 0.5
        if is_bot:
            print("  Bot aktivitesi ÅŸÃ¼phesi: Tekrarlayan tweetler tespit edildi.")
        return is_bot


    def _calculate_sentiment(self, tweets):
        """Basit sentiment analizi"""
        positive_words = ["good", "great", "amazing", "bullish", "buy", "long", "holding", "hodl", "strong"]
        negative_words = ["scam", "rug", "fake", "avoid", "dump", "short", "sell", "warning", "risk"]

        total_score = 0
        if not tweets:
            return 0.5 # Tweet yoksa nÃ¶tr

        for tweet in tweets:
            text = tweet.text.lower()
            pos_count = sum(word in text for word in positive_words)
            neg_count = sum(word in text for word in negative_words)

            # Sentiment skorunu hesapla (basit oran)
            score = 0.5 # Default nÃ¶tr
            if pos_count > neg_count:
                score = 1.0 # Pozitif
            elif neg_count > pos_count:
                score = 0.0 # Negatif
            # EÄŸer pos_count == neg_count ise skor 0.5 kalÄ±r (nÃ¶tr)

            total_score += score

        # Ortalama sentiment
        average_sentiment = total_score / len(tweets)
        return average_sentiment
```

```python
%%writefile utils/code_analysis.py
import re

class CodeAnalyzer:
    def __init__(self):
        self.risk_patterns = {
            "mint_function": [
                r"function\s+mint\s*\(",
                r"function\s+_mint\s*\(",
                r"\.mint\s*\(",
                r"mintTo\s*\("
            ],
            "ownership_issues": [
                r"onlyOwner",
                r"_owner\s*=",
                r"transferOwnership",
                r"renounceOwnership"
            ],
            "liquidity_removal": [
                r"removeLiquidity",
                r"withdraw.*Liquidity",
                r"emergencyWithdraw"
            ],
            "tax_functions": [
                r"buyTax\s*=",
                r"sellTax\s*=",
                r"setTax",
                r"updateTax"
            ],
            "blacklist_functions": [
                r"blacklist",
                r"addBot",
                r"removeBot",
                r"setBot"
            ],
            "pause_functions": [
                r"pause\s*\(",
                r"unpause\s*\(",
                r"setPaused"
            ]
        }

    def analyze_contract_code(self, address, source_code=""):
        """Kontrat kodunu analiz ederek risk bayraklarÄ±nÄ± dÃ¶ndÃ¼r"""
        flags = []

        if not source_code:
            flags.append("âš ï¸ Kontrat kodu doÄŸrulanamadÄ±")
            return flags

        source_code = source_code.lower()

        # Mint fonksiyonu kontrolÃ¼
        if self._check_pattern_exists(source_code, self.risk_patterns["mint_function"]):
            flags.append("ðŸ”´ Mint fonksiyonu tespit edildi")

        # Ownership kontrolleri
        if self._check_pattern_exists(source_code, self.risk_patterns["ownership_issues"]):
            if not self._check_ownership_renounced(source_code):
                flags.append("ðŸ”´ Owner haklarÄ± korunuyor")

        # Likidite Ã§ekme fonksiyonlarÄ±
        if self._check_pattern_exists(source_code, self.risk_patterns["liquidity_removal"]):
            flags.append("ðŸ”´ Likidite Ã§ekme fonksiyonu var")

        # Vergi fonksiyonlarÄ±
        if self._check_pattern_exists(source_code, self.risk_patterns["tax_functions"]):
            tax_rate = self._extract_tax_rate(source_code)
            if tax_rate > 10:
                flags.append(f"ðŸ”´ YÃ¼ksek vergi oranÄ±: %{tax_rate}")

        # Blacklist fonksiyonlarÄ±
        if self._check_pattern_exists(source_code, self.risk_patterns["blacklist_functions"]):
            flags.append("ðŸ”´ Blacklist fonksiyonu tespit edildi")

        # Pause fonksiyonlarÄ±
        if self._check_pattern_exists(source_code, self.risk_patterns["pause_functions"]):
            flags.append("ðŸ”´ Trading durdurma fonksiyonu var")

        # Honeypot kontrolleri
        if self._check_honeypot_patterns(source_code):
            flags.append("ðŸ”´ Honeypot kalÄ±plarÄ± tespit edildi")

        if not flags:
            flags.append("âœ… Kod analizi temiz")

        return flags

    def _check_pattern_exists(self, code, patterns):
        """Verilen pattern'lerden herhangi biri var mÄ± kontrol et"""
        for pattern in patterns:
            if re.search(pattern, code, re.IGNORECASE):
                return True
        return False

    def _check_ownership_renounced(self, code):
        """Ownership renounce edilmiÅŸ mi kontrol et"""
        renounce_patterns = [
            r"renounceownership\s*\(\s*\)",
            r"_owner\s*=.*0x0+",
            r"owner.*=.*address\(0\)"
        ]
        return self._check_pattern_exists(code, renounce_patterns)

    def _extract_tax_rate(self, code):
        """Kod iÃ§inden vergi oranÄ±nÄ± Ã§Ä±karma"""
        tax_patterns = [
            r"buytax\s*=\s*(\d+)",
            r"selltax\s*=\s*(\d+)",
            r"tax.*=\s*(\d+)"
        ]

        max_tax = 0
        for pattern in tax_patterns:
            matches = re.findall(pattern, code, re.IGNORECASE)
            for match in matches:
                max_tax = max(max_tax, int(match))

        return max_tax

    def _check_honeypot_patterns(self, code):
        """Honeypot kalÄ±plarÄ±nÄ± kontrol et"""
        honeypot_patterns = [
            r"require.*balances\[.*\]\s*>=",  # Balance kontrolÃ¼
            r"if.*balances\[.*\].*return false",  # Transfer engelleme
            r"mapping.*bool.*isexcluded",  # Exclusion mapping
            r"tradingopen\s*=\s*false"  # Trading kapalÄ±
        ]
        return self._check_pattern_exists(code, honeypot_patterns)
```

```python
%%writefile utils/feature_engineering.py
import numpy as np
from datetime import datetime

class FeatureEngineer:
    def build_features(self, onchain_data, social_data=None):
        """Ham verilerden ML modeli iÃ§in Ã¶zellikler oluÅŸtur"""

        # Temel Ã¶zellikler
        features = [
            onchain_data.get("holder_concentration", 75),
            int(onchain_data.get("mint_function", True)),
            int(onchain_data.get("lp_locked", False)),
            onchain_data.get("max_tax", 25),
            int(onchain_data.get("owner_renounced", False)),
            int(onchain_data.get("contract_verified", False))
        ]

        # Kontrat yaÅŸÄ± (gÃ¼n cinsinden)
        creation_date = onchain_data.get("creation_date", 0)
        if creation_date > 0:
            age_days = (datetime.now().timestamp() - creation_date) / 86400
            features.append(min(age_days, 365))  # Max 1 yÄ±l
        else:
            features.append(0)

        # Holder sayÄ±sÄ± (log Ã¶lÃ§eÄŸi)
        holder_count = onchain_data.get("holder_count", 1)
        features.append(np.log10(max(holder_count, 1)))

        # Sosyal medya Ã¶zellikleri
        if social_data:
            features.extend([
                social_data.get("twitter_mentions", 0),
                social_data.get("sentiment_score", 0.5) * 100,  # 0-100 Ã¶lÃ§eÄŸi
                int(social_data.get("bot_activity", False)),
                int(social_data.get("pump_dump_keywords", False))
            ])
        else:
            features.extend([0, 50, 0, 0])  # Default deÄŸerler

        return np.array(features)

    def get_feature_names(self):
        """Ã–zellik isimlerini dÃ¶ndÃ¼r"""
        return [
            "holder_concentration",
            "mint_function",
            "lp_locked",
            "max_tax",
            "owner_renounced",
            "contract_verified",
            "age_days",
            "log_holder_count",
            "twitter_mentions",
            "sentiment_score",
            "bot_activity",
            "pump_dump_keywords"
        ]
```
